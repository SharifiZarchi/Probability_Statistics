        میتوانید این سری را 1000 بار تکرار کنید و تعداد باری که تیم با احتمال برد p برنده میشود را گزارش کنید.

```r
library(ggplot2)
library(tidyverse)
sim_fn <- function(simN, n, p) {
    mean(rbinom(simN, n, p) > n / 2)
}
```

<div dir = "rtl" style="direction:rtl;line-height:250%;">
	<font face="XB Zar" size=5>
        <div dir = "rtl" style="direction:rtl;line-height:200%;">
	<font face="XB Zar" size=5>
        حال برای p های بزرگتر از نیم احتمال برنده شدن سری را محاسبه کنید.
        و نمودار مناسبی برای نمایش این اعداد رسم کنید.
</div>

```r
n <- 7
df <- data.frame(p = seq(0.5, 1, length.out = 26), n = n)

set.seed(1043)
simN <- 100000
df$win_prob <- apply(df, 1, function(x) sim_fn(simN, x[2], x[1]))
ggplot(df, aes(x = p, y = win_prob)) + geom_line() + 
                     labs(title = paste("Win probability for best of", n, "series"),
                          x = "Win prob for single game",
                          y = "Win prob for series")
```

        احتمال اینکه یک سری را ببازند کمتر میشود و این نسبت را ببینید، مثلا برای p = 0.6 مقادیر مختلف n را بررسی کنید.

```r
# simulation for multiple n values
p <- seq(0.5, 1, length.out = 26)
n <- seq(1, 15, length.out = 8)
df <- expand.grid(p, n)
names(df) <- c("p", "n")

set.seed(422)
simN <- 100000
df$win_prob <- apply(df, 1, function(x) sim_fn(simN, x[2], x[1]))
ggplot(df, aes(x = p, y = win_prob, col = factor(n))) +
    geom_line() +
    labs(title = paste("Win probability for best of n series"),
         x = "Win prob for single game", y = "Win prob for series") +
    theme(legend.position = "bottom")
```

            تابع قسمت قبل را به نحوی باز سازی کنید که هم احتمال برد در خانه و احتمال برد خارج از خانه و اینکه 4 بازی میکند یا 3 بازی را دریافت و مانند تابع قبلی احتمال برد

```r
sim_fn <- function(simN, n, home_p, away_p, start) {  
  if (start == "home") {
    more_p <- home_p
    less_p <- away_p
  } else {
    more_p <- away_p
    less_p <- home_p
  }
  
  mean(rbinom(simN, (n+1)/2, more_p) + rbinom(simN, (n-1)/2, less_p) > n / 2)
}
```

        برای رسم همه نمودار ها در یکجا درباره facet_wrap() مطالعه کنید.

```r
home_p <- seq(0, 1, length.out = 51)
away_p <- seq(0, 1, length.out = 51)
df <- expand.grid(home_p = home_p, away_p = away_p, start = c("home", "away"), stringsAsFactors = FALSE)

set.seed(111)
simN <- 5000
n <- 7
df$win_prob <- apply(df, 1, function(x) sim_fn(simN, n, as.numeric(x[1]), as.numeric(x[2]), x[3]))
ggplot(df, aes(x = home_p, y = win_prob, col = start)) +
  geom_line() +
  labs(title = paste("Win probability for best of", n, "series"),
       x = "Win prob for single game", y = "Win prob for series")+ facet_wrap(~ away_p)
```

        در یک فرایند پواسون زمان بین دو اتفاق از یک توزیع نمایی پیروی می‌کند. اگر بخواهیم زمان اتفاق n‌ام را پیش‌بینی کنیم از توزیع گاما استفاده می‌کنیم.
        $$T_n=\sum_{i=1}^n{X_i},\  X_i  s\ are\ i.i.d\ and\ X_i\sim Exp(\lambda)$$
        $$T_n\sim Gamma(n, \lambda)$$
       حال فرض کنید فاصله زمانی ورود دانشجویان به کلاس درس حضوری از توزیع پوآسون پیروی میکند.

        نمودار هیستوگرام شما باید چگالی احتمال را برای هر نمونه را نشان دهد.

```r
plot_hist_exp <- function(Ks, rate) {
data <- rexp(n = 1000, rate = rate)
probs <- dexp(x = Ks, rate = rate)
hist(data, probability = T, ylab = "P(X=k)", xlab = "k", col = 'steelblue', main = "")
points(Ks, probs, col = 'firebrick', pch=19)
legend("topright", legend=c("Observed data", "Exp dist"), lty = c(1, NA), pch = c(NA, 19), col = c('steelblue', 'firebrick', 'forestgreen'))
}

plot_hist_exp(c(1:1000), 1/5)
```

```r
library(ggplot2)
n <- 1000
lambda <- 1/5
shape <- 10

s <- rep(0, n)
for (i in 1:shape) {
  s <- s + rexp(n, lambda)
}

d_exp <- data.frame(x=s)
d_gamma <- data.frame(x=rgamma(n, shape=shape, rate=lambda))
ggplot() +
  geom_histogram(data=d_exp, aes(x), fill="red", alpha=0.5, bins = 50) +
  geom_histogram(data=d_gamma, aes(x), fill="blue", alpha=0.5, bins = 50)
```
