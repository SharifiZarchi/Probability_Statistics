    <br>
        یک گروه تحقیقاتی به تهران آمده و پس از ۱ دقیقه آن را ترک می‌کند (!!). وظیفه این گروه اندازه‌گیری مساحت تهران بوده است. این گروه پس از خروج خود از تهران مساحت آن را ۱۰۰ متر مربع اعلام می‌کند و می‌گوید این اندازه‌گیری با دقت بالای ۹۵ درصد کمتر از ۰.۱ خطا دارد.
    </br>
            می‌دانیم هر آزمایش دقیقا به یک ثانیه زمان نیاز دارد، در نتیجه این تیم ۶۰ بار آزمایش را انجام داده است.
    <br>
    </br>
        همچنین Yi مساحت اندازه گیری شده در دفعه iام است که برای آن داریم:
      <div dir="ltr">
      <p style="text-align:center;">
        Yi = مساحت دقیق تهران + Xi
       </p>
      </div>
        که Xi میزان خطای محاسبه در دفعه iام است. فرض می‌کنیم Xiها i.i.d هایی با 0 = E(Xi) و 4 = Var(X) هستند. گروه متوسط مساحت اندازه‌گیری شده را به صورت زیر بیان می‌کند:
            <div dir="ltr">
      <p style="text-align:center;">
        M = (Y1 + Y2 + ... + Yn) / n
       </p>
      </div>
      </br>
    سهراب پس از اعلام این نتایج خون جلوی چشمانش را می‌گیرد و به رسانه‌های ایرانی می‌گوید: "خورشید نمیمونه پشت ابر". او می‌خواهد نشان دهد تعداد آزمایش انجام شده بسیار کمتر از مقدار مورد نیاز است!
    <br>
    سهراب از شما می‌خواهد به او بگویید احتیاج به چند آزمایش است تا M با دقت بالای ۹۵ درصد کمتر از ۰.۱ با مساحت دقیق تهران اختلاف داشته باشد.
