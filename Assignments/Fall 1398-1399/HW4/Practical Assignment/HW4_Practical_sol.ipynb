```r
n = 100000
X <- runif(n, -1, 1)
Y <- sapply(X, function(x) runif(1, -sqrt(1-x^2), +sqrt(1-x^2)))
library(ggplot2)
ggplot(data.frame(X,Y), aes(x=X,y=Y, ymin = -1, ymax = 1, xmin = -1, xmax = 1), asp = 1) + geom_point(size = 0.1, color = 'blue', alpha  = 0.3) + coord_equal() 
hist(X)
hist(Y)    
```

```r
n = 100000
R <- runif(n, 0, 1)
Theta <- runif(n, -pi, pi)
library(ggplot2)
X = R*cos(Theta)
Y = R*sin(Theta)
ggplot(data.frame(X,Y), aes(x=X,y=Y, ymin = -1, ymax = 1, xmin = -1, xmax = 1), asp = 1) + geom_point(size = 0.1, color = 'blue', alpha  = 0.3) + coord_equal()
hist(X)
hist(Y)
```

```r
n = 100000
R <- sqrt(runif(n, 0, 1))
Theta <- runif(n, 0, 2*pi)
library(ggplot2)
X = R*cos(Theta)
Y = R*sin(Theta)
ggplot(data.frame(X,Y), aes(x=X,y=Y, ymin = -1, ymax = 1, xmin = -1, xmax = 1), asp = 1) + geom_point(size = 0.1, color = 'blue', alpha  = 0.3) + coord_equal()
```
