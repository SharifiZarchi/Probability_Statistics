<div id="sec_reminder" style="direction:rtl;line-height:300%;">

یادآوری
        </font>
		<p></p>
        <hr>
        واریانس در یک متغیر تصادفی را به یاد آورید. قصد داشتیم میزان تغییر یک متغیر تصادفی نسبت به میانگینش را به دست آوریم. برای این کار ابتدا اختلاف هر مشاهده با میانگین را حساب کردیم. با جمع زدن این اختلافات به مقدار صفر رسیدیم. برای رفع این مشکل مجذور اختلافات را با هم جمع کرده و میانگین گرفتیم.
        <br> <br>
        	فرمول واریانس به صورت زیر بود: 
            <br> 
            $$Var(X) = E[(X - \mu_X)^2] = E[X^2] - \mu_X^2$$
    </font>
</div>

<div id="sec_cov_meaning" style="direction:rtl;line-height:300%;">
    <br>
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
معنای کوواریانس
        </font>
		<p></p>
        <hr>
        با توجه به اینکه واژه‌ی co به معنای «هم» است، کوواریانس را با استفاده از معنی واریانس معنی می‌کنیم. کوواریانس میزان تغییر دو متغیر با همدیگر نسبت به میانگینشان است.
        	فرمول کوواریانس به صورت زیر است:
            <br><br>
            $$Cov(X,Y) = E[(X - \mu_X)(Y - \mu_Y)]$$
            <br>
            <font color=#2c7 size=5><b>تمرین.</b></font>
            سعی کنید اثبات کنید$$Cov(X, Y) = E[XY] - \mu_x.\mu_y$$
            هم‌چنین با کمی دقت متوجه می‌شوید که $$Cov(X, X) = Var(X)$$
	</font>
</div>

<div id="sec_cov_disc_cont" style="direction:rtl;line-height:300%;">
    <br>
	<font face="XB Zar" size=5>
        <font color=#FF7500 size=6>
 محاسبه‌ی کوواریانس در متغیرهای تصادفی گسسته و پیوسته
        </font>        
        <p></p>
        <hr>
        فهمیدیم که کوواریانس دو متغیر تصادفی، در واقع امیدریاضی تابعی از آن دو است. با در نظر گرفتن این نکته، به دو گزاره‌ی زیر توجه کنید:
        <br>
        <ul><li>
        اگر دو متغیر تصادفی گسسته‌ی $X$ و $Y$ دارای توزیع توأم $p(x,y)$ باشند آنگاه کوواریانس آن دو مطابق زیر محاسبه می‌شود:
            </li></ul>
        $$Cov(X,Y) = \sum_{i=1}^{n} \sum_{j=1}^{m} p(x_i, y_j)(x_i -\mu_X)(y_j - \mu_Y) = \Big(\sum_{i=1}^{n} \sum_{j=1}^{m} p(x_i, y_j)x_iy_j \Big)~-~ \mu_X\mu_Y$$
   <ul><li>
اگر دو متغیر تصادفی پیوسته‌ی $X$ و $Y$ بر روی بازه‌ی $[a, b] \times [c,d]$ دارای توزیع توأم $f(x, y)$ باشند آنگاه کوواریانس آن دو مطابق زیر محاسبه می‌شود: 
</ul></li>
    $$Cov(X, Y) = \int_c^d \int_a^b (x - \mu_x)\,(y - \mu_y)\,f(x,y) \,\mathrm{d}x\mathrm{d}y = \int_c^d \int_a^b x\,y\,f(x,y) \,\mathrm{d}x\mathrm{d}y - \mu_x\mu_y$$
   </font>
   <br>
</div>

<br>
<div id="sec_cov_disc_cont" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
        <font color=blue size=6>مثال.</font>
        <br>
        دو متغیر تصادفی X و Y به صورت زیر تعریف شده‌اند:
        $$X \sim U(-1,\,1)$$
        $$Y = X^2$$
        <br>
        می‌خواهیم کوواریانس این دو را به‌دست آوریم. در واقع باید عبارت زیر را محاسبه کنیم:
        <br>
        $$Cov(X, Y) = E[XY] - E[X]E[Y] = E[X.X^2] - E[X]E[X^2] = E[X^3] - E[X]E[X^2]$$
        <br>
         اگر $X$ یک متغیر تصادفی با توزیع یکنواخت در محدوده‌ی $[a, b]$ باشد داریم:
         $$E[X] = \int_a^bx.f(x)\,\mathrm{d}x = \int_a^bx.\frac{1}{b - a}\,\mathrm{d}x = \frac{x^2}{2\,(b - a)}\big|_{x = a}^{x = b} = \frac{b + a}{2}$$
         <br>
         $$E[X^2] = \int_a^bx^2.f(x)\,\mathrm{d}x = \int_a^bx^2.\frac{1}{b - a}\,\mathrm{d}x = \frac{x^3}{3\,(b - a)}\big|_{x = a}^{x = b} = \frac{b^3 - a^3}{3\,(b - a)}$$
         <br>
         $$E[X^3] = \int_a^bx^3.f(x)\,\mathrm{d}x = \int_a^bx^3.\frac{1}{b - a}\,\mathrm{d}x = \frac{x^4}{4\,(b - a)}\big|_{x = a}^{x = b} = \frac{b^4 - a^4}{4\,(b - a)}$$
         <br> 
         با توجه به روابط بالا می‌توانیم عبارات زیر را بدست آوریم:
         <br>
         $$E[X] = 0 \quad \quad \quad  E[X^2] = \frac{1}{3} \quad  \quad \quad E[X^3] = 0$$
         <br>
         با جایگذاری در رابطه‌ی اول می‌رسیم به اینکه $Cov(X,Y) = 0$
         <br><br>         پس دیدیم با اینکه متغیر تصادفی های $X$ و $Y$ به هم وابسته بودند، حاصل کوواریانس آنها صفر شد. این را در R نیز تست می‌کنیم. با استفاده از دستور runif مشاهده‌هایی از متغیر تصادفی $X$ و سپس با استفاده از آنها مشاهده‌هایی از $Y$ را تولید می‌کنیم. در نهایت با استفاده از دستور cov کوواریانس آن دو را بدست می‌آوریم:
</div>

```r
X <- runif(n = 10000, min = -1, max = 1)
Y <- X ^ 2
cov(X, Y)
```

<div id="sec_cov_attr" style="direction:rtl;line-height:300%;">
    <br>
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>ویژگی‌های کوواریانس</font>
		<p></p>
        <hr>
        چند مورد از ویژگی‌های کوواریانس را در زیر آورده‌ایم:
        <br><br>
        $1. \quad Cov(X_1 + X_2\,,Y) = Cov(X_1\,,Y) + Cov(X_2\,,Y) \\
        2. \quad Cov(aX + b\,,cY + d) = acCov(X\,,Y) \\
        3. \quad Var(X + Y) = Var(X) + Var(Y) + 2Cov(X\,,Y) \\
        \quad 3.1. \quad Var\Big(\sum_{i = 1}^nX_i\Big) = \sum_{i = 1}^n\sum_{j = 1}^nCov(X_i\,,X_j) =   
\sum_{i = 1}^nVar\big(X_i\big) + 2\sum_{(i, j): i < j}Cov(X_i\,, X_j) \\
        4. \quad if~X~and~Y~are~independent~then~Cov(X\,,Y) = 0$
        <br><br>
		<font color=red size=6><b>تذکر مهم.</b></font>
        دقت کنید که برعکس مورد ۴ صادق نیست. یعنی صرفا صفر بودن کوواریانس دو متغیر تصادفی بیانگر استقلال آنها از هم نیست.
        <br>
        <font color=red size=6>تذکر.</font>
        با استفاده از موارد ۳ و ۴ میتوان دید که اگر دو متغیر تصادفی مستقل از هم باشند، واریانس جمع‌شان برابر با جمع واریانس‌هایشان می‌باشد. 
        <br> <br>
        مورد ۲ را اثبات می‌کنیم:  
        <br>
        $Cov(aX + b\,, cY + d) = E[(aX + b)(cY + d)] - E[aX + b]E[cY + d] = E[acXY + adX + bcY + bd] - (aE[X] + b)(cE[Y] + d) = acE[XY] + adE[X] + bcE[Y] + bd - acE[X]E[Y] - adE[X] - bcE[Y] - bd = acE[XY] - acE[X]E[Y] = ac(E[XY] - E[X]E[Y]) = acCov(X,Y)
$
        <br> <br>
	</font>
</div>

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
        <font color=#2c7 size=5><b>تمرین ۱.</b></font>
        سعی کنید سایر ویژگی‌ها را نیز به صورت تئوری اثبات کنید.
        <br>
        <font color=#2c7 size=5><b>تمرین ۲.</b></font>
        ثابت کنید که در صورتی‌ که $X$ و $Y$ دو متغیر تصادفی و $n$ عددی ثابت باشند اگر داشته باشیم $X + Y = n$ آن‌گاه $Cov(X\,,Y) = -Var(X)$
        <br>
        راهنمایی.  از ویژگی ۲ استفاده کنید.
        <br>
        <font color=#2c7 size=5><b>تمرین ۳.</b></font>
        اگر $X$ و $Y$ دو متغیر تصادفی با واریانس برابر باشند، ثابت کنید $X + Y$ و $X - Y$ دارای کوواریانس صفر هستند. 
    </font>
</div>


<div id="sec_example" style="direction:rtl;line-height:300%;">
    <br>
	<font face="XB Zar" color=#FF7500 size=6>
مثال: مرد خوشحال
	</font>
    <hr>
	<font face="XB Zar" size=5>
    مرد خوشحالی روی مرکز صفحه‌ی مختصات (0,0) ایستاده و در هر مرحله یک واحد با احتمال برابر به یکی از جهت‌های شمال، جنوب، شرق یا غرب حرکت می‌کند.
    <br>
    متغیر تصادفی‌های $(X_n,Y_n)$ و $R_n$ را به ترتیب موقعیت او در صفحه‌ی مختصات و فاصله‌ی او از مرکز مختصات پس از $n$ مرحله در نظر بگیرید.
    <br>
    الف. تعیین کنید که آیا $X_n$ و $Y_n$ از هم مستقل هستند یا خیر؟
    <br>
      ب. $Cov(X_n,Y_n)$ را بیابید.
        <br>
    پ. $E[R_n^2]$ را بیابید.
        <br>
    <br>
    <font color=blue size=5>
    پاسخ.
    </font>
        برای اینکه نشان دهیم $X_n$ و $Y_n$ از هم مستقل نیستند، یک مثال حداکثری میزنیم. اگر فرض کنیم مرد خوشحال در همه‌ی مراحل به سمت شرق حرکت کرده باشد، می‌توان نتیجه گرفت حتما $Y_n=0$ است. این یعنی $Pr(Y_n = 0|X_n = n) = 1$.
    <br><br>
    توجه کنید $X_n$ و $Y_n$ را می‌توان به صورت زیر نوشت:
    $$X_n = \sum_{i=1}^n H_i \quad , \quad Y_n = \sum_{j=1}^n V_j$$
    به طوری که $H_i$ جهت حرکت در مرحله‌ی $i$اُم روی محور xها را نشان می‌دهد؛ حرکت به سمت غرب $-1$، حرکت به سمت شرق $1$ و در غیر اینصورت $0$ است. $V_j$ به طور مشابه حرکت روی محور yها را نشان می‌دهد. طبق ویژگی‌های کوواریانس می‌توان $Cov(X_n,Y_n)$ را به صورت زیر نوشت:
    $$Cov(\sum_{i=1}^n H_i , \sum_{j=1}^n V_j) = \sum_{i=1}^n \sum_{j=1}^n Cov(H_i,V_j)$$
    در حالتی که $i \neq j$، $H_i$ و $V_j$ از هم مستقل‌اند و حاصل کوواریانسشان صفر است. اما $H_i$ و $V_i$ به هم شدیدا وابسته‌اند. در واقع حتما یکی از آنها صفر است، یعنی $Cov(H_i,V_i) = E[H_iV_i] - E[H_i]E[V_i] = 0$. پس:
    $$Cov(X_n,Y_n) = \sum_{i,j}Cov(H_i,V_j) = 0.$$
    از آنجایی که برای $i \neq j$، $H_i$ و $H_j$ از هم مستقل‌اند پس کوواریانس آنها صفر است:
    $$Cov(H_i,H_j) = 0 \quad \rightarrow \quad E[H_iH_j] - E[H_i]E[H_j] = 0 \quad \rightarrow \quad E[H_iH_j] = 0$$
    با داشتن روابط $R_n^2 = X_n^2 + Y_n^2$ و $E[H_iH_j] = 0$ برای $i \neq j$ و کمی تقارن، نتیجه می‌گیریم:
    $$E[R_n^2] = E[X_n^2] + E[Y_n^2] = 2E[X_n^2] = 2nE[H_1^2] = n$$
    </font>
</div>

<div id="sec_corr_meaning" style="direction:rtl;line-height:300%;">
    <br>
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
معنای هم‌بستگی
        </font>
		<p></p>
        <hr>
        در قسمت قبل که از کوواریانس حرف زدیم، صرفا به علامت آن توجه کردیم و نه به مقدار آن. در واقع مقدار آن هیچ اطلاعاتی به ما نمی‌دهد.
        <br>
        واحد کوواریانس، واحد متغیر $X$ در واحد متغیر $Y$ است. از روی فرمول $Cov(X,Y) = E[XY] - \mu_X \mu_Y$ هم می‌توان این را حدس زد. به همین دلیل اگر ما صرفا واحد‌های متغیر‌ها را تغییر بدهیم، عدد کوواریانس نیز تغییر خواهد کرد. 
<br>
مثلا نسبت قد و وزن انسان‌ها را در نظر بگیرید. فرض کنید یک بار قد را بر حسب سانتی‌متر و یک بار بر حسب متر نوشته‌ایم. در حالت اول کوواریانس صد برابر حالت دوم خواهد شد، در حالی که تغییر قد و وزن نسبت به همدیگر تفاوتی نکرده‌است.
        <br>
        برای رفع این مشکل هم‌بستگی را تعریف می‌کنیم.
        <br><br>
        متغیر $X$ و $Y$ را استاندارد میکنیم:
    $$X \rightarrow \frac{X - \mu_X}{\sigma_X} \quad \quad Y \rightarrow \frac{Y - \mu_Y}{\sigma_Y} $$
    <br>
    حال کوواریانس این دو متغیر استاندارد‌شده را در نظر بگیرید. (دقت کنید که میانگین صفر شده است):
    <br> <br>
$$ Cov(\frac{X - \mu_X}{\sigma_X}, \frac{Y - \mu_Y}{\sigma_Y}) = 
    E[\frac{X - \mu_X}{\sigma_X} \frac{Y - \mu_Y}{\sigma_Y}] = 
\frac{1}{\sigma_X\sigma_Y}E[(X - \mu_X)(Y - \mu_Y)] = 
\frac{Cov(X, Y)}{\sigma_X\sigma_Y}$$
<br>
واضح است که در کسر نهایی، واحد صورت و مخرج یکسان است. پس مشکل مقیاس که بالاتر مطرح کردیم را نداریم. 
<br>
مقدار $\frac{Cov(X,~Y)}{\sigma_X\sigma_Y}$ را هم‌بستگی یا $\rho$ می‌نامیم. 
</font>
</div>

<div id="sec_corr_attr" style="direction:rtl;line-height:300%;">
    <br>
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
خواص هم‌بستگی
        </font>
        <hr>
        <br>
    بعضی از خواص هم‌بستگی عبارتند از:
    <ul>
    <li> هم‌بستگی مقیاس ندارد. یعنی اگر مقداری را در $X$ یا $Y$ ضرب کنیم، مقدار هم‌بستگی تغییر نخواهد کرد.</li>
    <li> هم‌بستگی دو متغیر تصادفی مستقل از هم برابر با صفر است. </li>
    <li> $ -1 \leq \rho \leq 1$</li>
    </ul>
    خاصیت آخر مهم‌ترین خاصیت هم‌بستگی است. هرچقدر مقدار هم‌بستگی به $1$ یا $-1$ نزدیک‌تر باشد، میزان رابطه‌ی خطی $X$ و $Y$ نیز (مثبت یا منفی) افزایش می‌یابد.
    همچنین داریم:
    <br> <br>
    $$ 
    \rho = +1  \iff  Y = aX + b, ~ a > 0 \\
    \rho = -1 \iff Y = aX + b, ~ a < 0
    $$
    <br>
   	</font> 
</div>

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
        <font color=#2c7 size=5><b>تمرین ۱.</b></font> 
            سعی کنید دو رابطه‌ی بالا را اثبات کنید.
            <br>
            راهنمایی: عبارات $Var(\frac{X}{\sigma_X} \pm \frac{Y}{\sigma_Y})$ 
            را در نظر بگیرید.
            <br> 
        <font color=#2c7 size=5><b>تمرین ۲.</b></font>
        عبارت $Cov(X,Y)^2 \le Var(X)Var(Y)$ یکی از صورت‌های نامساوی کوشی-شوارتز است. با استفاده از آن ثابت کنید هم‌بستگی همواره بین $-1$ و $1$ است.
</font>
</div>

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
        در زیر، نمودار دو متغیر تصادفی با هم‌بستگی دلخواه را مشاهده میکنید. برای تغییر مقدار هم‌بستگی، خط سوم را تغییر دهید. همچنین مفید است که بدانید هم‌بستگی بزرگتر از $0.7$ یا کوچکتر از $-0.7$، هم‌بستگی قوی محسوب می‌شود.
</font>
</div>

```r
library(ggplot2)

corr <- 0.7 # change this parameter
points <- 4000
x <- rnorm(points, 0, 1)
n <- rnorm(points, 0, 1)
a <- if (corr) (sqrt((1 - corr^2)/(corr^2)) - 1)/((1-2*corr^2)/(corr^2)) else 0
b <- 1 - a
y <- (if (corr < 0) -1 else 1) * (a*x+b*n)
ggplot() +
  geom_point(aes(x = x, y = y), size = 0.6, col = "blue", alpha = 0.6) +
  coord_cartesian(xlim = c(-4, 4), ylim = c(-4, 4)) +
  theme_minimal()
```

هم‌بستگی و علیت
        </font>
        <hr>
        <br>
حال که معنای هم‌بستگی را یاد گرفتید شاید به ذهنتان برسد که اگر دو متغیر $X$ و $Y$ هم‌بستگی مثبت داشته باشند، یعنی افزایش یکی <b>موجب</b> افزایش دیگری می‌شود. 
ولی این تفکر درستی نیست.
<br> <br>
<center><b>هم‌بستگی به معنای علیت نیست</b></center>
<br>
مثلا فرض کنید $X$ متغیر تصادفی تعداد بستنی‌های فروخته شده و $Y$ تعداد حمله‌های کوسه‌ها در هر ماه باشد. اگر هم‌بستگی این دو مثبت باشد، دلیل بر این نمی‌شود که فروش زیاد بستنی‌ها موجب حمله‌ی کوسه بشود. 
احتمالا فروش زیاد بستنی به علت گرمی هوا بوده، و در هوای گرم تعداد افراد کنار ساحل و داخل دریا بیشتر می‌شود و بدین ترتیب حمله‌ي کوسه‌ها نیز افزایش می‌یابد.
<br>
