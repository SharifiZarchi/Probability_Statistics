<div id="sec_intro" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مقدمه
        </font>
        <p></p>
        <hr>
        اکثر مسائلی که تا اینجای کار بررسی کردیم به این شکل بودند که ما ویژگی های یک جامعه بزرگ (به عنوان مثال توزیع احتمال یک متغیر تصادفی در آن جامعه) را می‌دانستیم و می‌خواستیم براساس آن یک رویداد را پیش‌بینی کنیم. اما اگر بخواهیم ویژگی های جامعه را به دست آوریم باید چه کنیم؟ برای مثال اگر بخواهیم پیش‌بینی کنیم در انتخابات آتی چه نامزدی رای خواهد آورد باید نظر تمام افراد جامعه را بپرسیم تا بدانیم اکثریت جامعه به کدام نامزد رای خواهد داد اما 
        اینکار بسیار هزینه بر است. آیا راه حل دیگری وجود دارد؟
        <br><br>
        <b>استدلال استنتاجی در مقابل استدلال استقرایی</b><br>
        در منطق دو شیوه برای استدلال وجود دارد.
        <ul>
            <li>
        شیوه اول که به آن استدلال استنتاجی (deductive reasoning) می‌گویند یک شیوه رسیدن از کل به جز است که در آن با استفاده از حقایق و کلیاتی از که قبل می‌دانیم به نتایج جدید می‌رسیم.
            </li>
            <li>
                شیوه دوم استدلال استقرایی (inductive reasoning) است که یک شیوه رسیدن از جز به کل است و با مشاهده نتایج جزئی تلاش می‌کنیم آن‌ها را به کل تعمیم دهیم.
            </li>
        </ul>
        <img src="figs/prob_vs_stat.jpg" style="float:center;"/>
حال که با دو شیوه استدلال آشنا شدیم می‌توان گفت که علم احتمال نمونه‌ای از استدلال استنتاجی و آمار نمونه‌ای از استدلال استقرایی است. در مبحث احتمال دیدیم که می‌توان ویژگی‌های یک نمونه کوچک را از روی ویژگی‌های جامعه تعیین کرد. علم آمار به ما می‌گوید عکس این هم امکان‌پذیر است. یعنی می‌توان از روی ویژگی‌های یک نمونه کوچک از جامعه به ویژگی های کل جامعه پی‌برد. برای مثال اگر بخواهیم نتیجه انتخابات را پیش‌بینی کنیم می‌توانیم صرفا از یک نمونه ۱۰۰ یا ۱۰۰۰ نفری استفاده کنیم و اکثریت آن را به جامعه نیز تعمیم دهیم. البته این‌کار به شرطی امکان پذیر است که نمونه ما به اندازه کافی تصادفی باشد که آن را در ادامه بررسی می‌کنیم.
        <br>
        در نهایت می‌توان گفت که آمار و احتمال مکمل همدیگر هستند. در یک شیوه علمی ما ابتدا با جمع آوری نمونه‌ها برای کلیت جامعه مدل‌های آماری می‌سازیم و سپس با کمک احتمال و از روی مدل‌های ساخته شده به سوال‌های خود درباره جامعه پاسخ می‌دهیم.
    </font>
</div>

<div id="sec_infer" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
استنباط آماری
        </font>
        <p></p>
        <hr>
        به طور کلی استنباط آماری به دو شیوه تقسیم بندی می‌شود. شیوه اول که به آن شیوه <b> فروان‌گرایانه </b> (frequentist) می‌گویند فرض می‌کند پارامتری از جامعه که به دنبال تخمین مقدار آن هستیم یک مجهول غیر تصادفیست که از طریق نمونه‌گیری به تعداد زیاد قابل تخمین است. به عنوان مثال با این شیوه اگر بخواهیم میانگین یک جامعه را تخمین بزنیم یک نمونه تصادفی از افراد آن جامعه انتخاب می‌کنیم. جواب‌های آن ها را یادداشت میکنیم و میانگین آن‌ها را به عنوان میانگین جامعه انتخاب می‌کنیم.
        <br>
        استفاده از بازه اطمینان برای تخمین بازه‌ای و p-value برای آزمون فرض از مثال‌های به‌کارگیری این شیوه هستند که در ادامه با آن‌ها آشنا خواهید شد.
        <br><br>
        شیوه دوم که به آن شیوه <b> بیزی </b> (Bayesian) می‌گویند فرض می‌کند پارامتر موردنظر ما خود یک متغیر تصادفیست. بنابراین در این شیوه بحث درباره تخمین مقدار پارامتر نیست بلکه درباره تخمین توزیع آن است. در این شیوه ابتدا یک حدس اولیه درباره توزیع متغیر تصادفی موردنظرمان می‌زنیم و سپس با آزمایش و مشاهده یک نمونه، توزیع جدید آن را با کمک قانون بیز به دست می‌آوریم. به عبارت دیگر اگر متغیر تصادفی مورد علاقه ما $\theta$ باشد و متغیر تصادفی مورد آزمایش $X$ به $\theta$ بستگی داشته باشد آنگاه داریم 
        <br>
        <center>$ P(\theta|X) = \frac{P(X|\theta)P(\theta)}{P(X)}$</center>
        که در رابطه فوق $P(\theta)$ حدس اولیه ما و $P(\theta|X)$ نتیحه‌گیری ما از روی آزمایش است. تخمین‌گر MAP و توزیع conjugate prior از مثال‌های استفاده از این شیوه هستند. 
    </font>
</div>

کاربردهای آمار
        </font>
        <p></p>
        <hr>
        برای انجام تحلیل‌های آماری در مرحله اول به طراحی آزمایش و جمع‌آوری داده نیاز داریم. در این مرحله باید داده‌های ما به گونه‌ای مناسب برداشته شوند که آن‌را در ادامه بررسی می‌کنیم. اما سوال اینجاست که پس از جمع آوری این اطلاعات چه کارهایی می‌توان با آن‌ها انجام داد؟ 
        <ul>
            <li>
                فرض‌کنید طبق معمول یک سکه داریم که این بار احتمال شیرآمدنش را نداریم. برای اینکه بتوانیم این احتمال را حدس بزنیم سکه را چندین بار پرتاب می‌کنیم و تعداد شیرها را ثبت می‌کنیم. پس از انجام نمونه‌گیری به این شیوه، باید یک روش مناسب برای حدس زدن مجهولمان داشته باشیم که به آن <i><b> تخمین </b></i> می‌گویند. تخمین می‌تواند به صورت نقطه‌ای یا بازه‌ای باشد یا در آن مجهول ما یک متغیر تصادفی فرض شود یا صرفا یک پارامتر نامعلوم. در ادامه درس با انواع این تخمین ها آشنا خواهیم شد.
            <img src="figs/Point-Estimate-Interval-Estimate.jpg" style="float:center; width: 60%;">
                <center><figcaption style="font-size:60%;">تفاوت بین تخمین نقطه‌ای و بازه‌ای</figcaption></center>
            </li>
            <br/>
            <li>
                گاهی اوقات پس از جمع آوری داده‌ها می‌خواهیم درستی یا نادرستی یک فرض درباره جامعه را بررسی کنیم. مثلا پس از جمع آوری وزن افراد یک نمونه می‌خواهیم بررسی کنیم آیا وزن افراد جامعه از توزیع نرمال پیروی می‌کند یا خیر؟ برای این‌کار نیاز به ابزاری داریم تا به ما در تصمیم گیری اینکه فرضمان درست است یا خیر کمک کند. نام این ابزار <i><b> آزمون فرض </b></i> است و در ادامه با انواع آن و آزمون‌های مختلف آشنا می‌شویم. 
            </li>
            <img src="figs/hypothesis_testing.jpeg">
            <center><figcaption style="font-size:60%;">پذیرفتن یا رد کردن فرض صفر با کمک آماره (با این مفاهیم در ادامه آشنا خواهید شد)</figcaption></center>
            <br/>
            <li>
                اگر تا به حال از خود پرسیده باشید قوانین فیزیک چگونه کشف می‌شوند باید بگوییم به کمک آمار! بخشی از آمار به ما کمک می‌کند تا بتوانیم به کمک آزمایش های متعدد روابط بین متغیر‌ها را کشف کنیم که به این بخش از آمار <i><b> تحلیل رگرسیون </b></i> می‌گویند. در بخش پایانی درس با این بخش و به ویژه رگرسیون خطی آشنا خواهیم شد.
            </li>
            <img src="figs/linear_reg.png" style="float:center; wdith: 200%">
            <center><figcaption style="font-size:60%;">برازش خط به داده‌ها با رگرسیون خطی</figcaption></center>
        </ul>
        این موارد چند نمونه محدود از تحلیل‌های متنوعی بودند که می‌توانیم روی داده‌های آزمایش انجام دهیم و نتایج جالبی از آن به دست اوریم.
    </font>
</div>

<div id="sec_sample" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
نمونه‌گیری تصادفی
        </font>
        <p></p>
        <hr>
        انتخاب یک نمونه تصادفی اولین قدم در تحلیل‌های آماری است بنابراین می‌خواهیم کمی دقیق‌تر آن را بررسی کنیم. به طور کلی یک نمونه تصادفی یک دنباله 
        $$X_1, X_2, ..., X_n$$ از متغیر‌های تصادفی با توزیع یکسان است. اما در واقعیت ما نه با یک توزیع احتمال بلکه با یک جامعه سر و کار داریم. برای مثال فرض کنید می‌خواهیم قد
        افراد جامعه را بررسی کنیم. در اینجا می‌توان قد یک فرد را یک متغیر تصادفی مانند $X$ فرض کرد که توزیع احتمال آن از فراوانی نسبی قد افراد جامعه به‌دست می‌آید.
        <br>
        برای نمونه‌گیری از این جامعه باید $n$ نفر را انتخاب و قد آن‌ها را یادداشت کنیم اما نحوه انتخاب این $n$ نفر بسیار مهم است. برای مثال اگر در این جامعه بزرگ تنها دو نفر را انتخاب کرده و برای هربار یادداشت یک نمونه جدید بین این دو نفر شیر یا خط کنیم به جای اینکه $X_i$ها از توزیع جامعه پیروی کنند از یک توزیع برنولی پیروی خواهند کرد! برای اطمینان از اینکه $X_i$ها از توزیع جامعه پیروی می‌کنند (و در نتیجه نمونه ما به اندازه کافی تصادفی است) باید مطمئن شویم <b> هر کدام از اعضای جامعه شانس برابری برای حضور در نمونه دارند</b>. این نکته اهمیت بسیار زیادی دارد و گاهی فراموش می‌شود. برای مثال در گذشته که اینترنت فراگیر نبود نظرسنجی‌های اینترنتی راه خوبی برای نمونه‌گیری از جامعه نبودند!
        <br>
        همچنین در حالت کلی نباید فرض کرد $X_i$ها از هم مستقلند. در صورتی که تعداد اعضای جامعه نامتناهی باشد یک نمونه‌گیری تصادفی به استقلال $X_i$ها منجر خواهد شد اما در یک جامعه متناهی دو استراتژی برای نمونه‌گیری داریم:
        <ul>
            <li>
                استراتژی اول انتخاب با جایگذاری است که باعث استقلال $X_i$ها خواهد شد.
            </li>
            <li>
                استراتژی دوم انتخاب بدون جایگذاری است که در این صورت نمی‌توان استقلال $X_i$ها را تضمین کرد. هرچند اگر اندازه جامعه به قدر کافی بزرگ باشد می‌توان آن را نامتناهی و نمونه‌ها را مستقل در نظر گرفت.
            </li>
        </ul> 
    </font>
</div>

<div id="sec_statistic" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
آماره‌های معروف
        </font>
        <p></p>
        <hr>
        حالا که نمونه‌گیری را انجام دادیم زمان آن رسیده از آن اطلاعاتی را برداشت کنیم. به هر تابعی که روی یک نمونه تصادفی اعمال شود یک <b>آماره</b> (statistic) گفته می‌شود. با توجه به اینکه آماره تابعی از تعدادی متغیر تصادفی است خود نیز یک متغیر تصادفی است. در مثال‌های زیر فرض می‌کنیم $E(X) = \mu$ و $Var(X) = \sigma^2$ 
        <br/>
        <ul>
            <li>
                اولین و پرکاربردترین آماره معروف میانگین نمونه است که به شکل زیر تعریف می‌شود.
                $$\bar{X} = \frac{1}{n}\sum_{i=1}^{n}X_i$$
                امید ریاضی میانگین نمونه نیز به شکل زیر محاسبه می‌شود.
                $$\mathbb{E}(\bar{X}) = \frac{1}{n}\sum_{i=1}^{n}\mathbb{E}(X_i) = \frac{1}{n}n\mu$$
                و واریانس آن به شرط مستقل بودن نمونه‌ها
                $$Var(\bar{X}) = \frac{1}{n^2}\sum_{i=1}^{n}Var(X_i) = \frac{1}{n^2}n\sigma^2 = \frac{1}{n}\sigma^2$$
            </li>
            <li>
            از دیگر آماره‌های مهمی که میزان تغییرات در نمونه را نشان می‌دهد واریانس نمونه است. با توجه به تعریف واریانس برای یک متغیر تصادفی اولین تعریفی که برای واریانس نمونه به نظر می‌رسد تعریف زیر است.
            $$\frac{1}{n}\sum_{i=1}^{n}(X_i - \bar{X})^2$$
            اما در بخش‌های بعدی خواهید دید که اگر از این تعریف برای تخمین واریانس جامعه استفاده کنیم نتایج ما به سمت ضریب خاصی از واریانس جامعه متمایل خواهد شد (یا به عبارتی امید ریاضی آماره فوق با $\sigma^2$ برابر نیست) بنابراین از تعریف دیگری برای واریانس نمونه استفاده می‌کنیم که خاصیت فوق را داشته باشد.
            $$s^2 = \frac{1}{n-1}\sum_{i=1}^{n}(X_i - \bar{X})^2$$
            </li>
        </ul>
    </font>
</div>

مثال
        </font>
        <p></p>
        <hr> 
        حال می‌خواهیم به صورت تجربی رفتار میانگین و واریانس نمونه را به عنوان یک متغیر تصادفی بررسی کنیم. در قطعه کد زیر میانگین سایز برنامه‌های پلی استور گوگل (که داده‌های آن در فایل googleplaystore.csv آمده است) را بررسی می‌کنیم. هر کدام از نقاط کشیده شده در نمودارها میانگین یا انحراف معیار یک نمونه تصادفی هستند. با کمک نمودارهای زیر می‌توان کاهش تعییرات چه در میانگین و چه در انحراف معیار نمونه را با افزایش اندازه نمونه مشاهده کرد.
    </font>

```r
library(stringr)
library(ggplot2)


df <- read.csv("data/googleplaystore.csv")
sizes <- as.numeric(str_replace_all(df$Size, "M", ""))
sizes <- sizes[which(!is.na(sizes))]

max_sample_size <- 1000

sample_size <- c(1:max_sample_size, 1:max_sample_size)

means <- sapply(sample_size[1:max_sample_size], function (i) mean(sample(sizes, i)))
sds <- sapply(sample_size[(max_sample_size + 1):(2 * max_sample_size)], function (i) sd(sample(sizes, i)))

statistics <- c(means, sds)

types <- c(rep('mean', max_sample_size), rep('sd', max_sample_size))
real_values <- c(rep(mean(sizes), max_sample_size), rep(sd(sizes), max_sample_size))

data <- data.frame(x = sample_size, y = statistics, types = types, population = real_values)

ggplot(data, aes(x = x, y = y)) + geom_point(aes(color = 'samples')) + 
  geom_hline(aes(yintercept = population, color = 'population')) + 
  facet_grid(types~.) + 
  labs(x = 'size', y = 'statistic')
```

خلاصه
        </font>
        <p></p>
        <hr>
        <ul>
        <li>
            تقاوت آمار و احتمال:<br>
            احتمال از اطلاعات از پیش دانسته درباره جمعیت برای تخمین ویژگی‌های یک نمونه استفاده می‌کند. آمار از اطلاعات کسب شده از یک نمونه تلاش می‌کند ویژگی‌های جمعیت را به دست آورد.
        </li>
        <li>
            کاربردهای آمار:<br>
            علم آمار به ما کمک می‌کند کمیت‌های مورد نیازمان را تخمین بزنیم، درستی فرض‌های خود را بیازماییم و روابط پنهان بین متغیرها را کشف کنیم.
        </li>
        <li>
            نمونه‌گیری تصادفی:<br>
            یک نمونه‌گیری تصادفی است اگر شانس حضور هرکدام از اعضای جامعه در آن نمونه برابر باشد. 
        </li>
        <li>
            میانگین و واریانس نمونه:<br>
            آماره‌ها توابعی هستند که برروی نمونه‌های تصادفی اعمال می‌شوند و اطلاعات مفیدی درباره نمونه در اختیار ما می‌گدرند. از معروف ترین آماره‌ها میانگین و واریانس نمونه هستند.
            $$\bar{X} = \frac{1}{n}\sum_{i=1}^{n}X_i \quad s^2 = \frac{1}{n-1}\sum_{i=1}^{n}(X_i - \bar{X})^2$$
            که از ویژگی‌های مهم آن‌ها عبارت است از
            $$\mathbb{E}(\bar{X}) = \mathbb{E}(X) \quad \mathbb{E}(s^2) = Var(X)$$
            و در شرایطی که $X_i$ها مستقل باشند 
            $$Var(\bar{X}) = \frac{Var(X)}{n}$$
        </li>
        </ul>

<div id="sec_practice" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
تمرین
        </font>
        <p></p>
        <hr>
        گفتیم اگر نمونه‌گیری ما به صورت بدون جایگذاری از مجموعه متناهی باشد دیگر متغیرهای تصادفی ما مستقل نیستند. ثابت کنید در این حالت رابطه زیر برقرار است.
        $$Var(\bar{X}) = \frac{\sigma^2}{n}\frac{N - n}{N - 1}$$
        <br/>
        همانطور که از این روابط مشخص است واریانس میانگین نمونه در صورتی که نمونه‌گیری به صورت بدون جایگذاری باشد با ضریب $\frac{N - n}{N - 1}$ کاهش می‌یابد و در صورتی که با $n$ ثابت $N$ به بی‌نهایت میل کند این ضریب برابر با ۱ شده و بین نمونه‌گیری بدون جایگذاری و با جایگذاری تفاوتی نخواهد بود.
        <br/>
        (راهنمایی: ابتدا نشان دهید $Cov(X_i, X_j) = -\frac{\sigma^2}{N - 1}$ )
