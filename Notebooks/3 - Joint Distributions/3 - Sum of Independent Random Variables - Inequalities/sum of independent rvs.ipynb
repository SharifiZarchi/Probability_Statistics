<div id="continuous" style="direction:rtl; line-height:300%;">
    <div align=right>
        <font face="XB Zar" size=6 color=#FF7500>
            بررسی جمع دو متغیر تصادفی مستقل در حالت پیوسته
        </font>
        <p></p>
        <hr/>
        <font face="XB Zar" size=5>
                در مطالعه آمار و احتمالات، معمولا محاسبه توزیع X + Y از روی توزیع‌های X و Y زمانی که X و Y مستقل باشند، از اهمیت بالایی برخوردار است.
        فرض کنید X و Y دو متغیر تصادفی مستقل با توزیع‎های پیوسته $ f_X $ و $ f_Y $ باشند.
            تابع توزیع  X + Y را می‎توان به شکل زیر محاسبه کرد.
        <br/>
            ابتدا تابع توزیع تجمعی
            $
            F_{X+Y}
            $
            را محاسبه می‎کنیم:
       
$$
\begin{align*}
F_{X+Y}(a) &= P(X + Y \le a) \\ &= \iint\limits_{x + y \le a} f_X(x)f_Y(y) \, \mathrm{d}x \mathrm{d}y \\
&= \int_{- \infty}^{\infty} \int_{- \infty}^{a – y} f_X(x)f_Y(y) \, \mathrm{d}x \mathrm{d}y \\
&= \int_{-\infty}^{\infty} f_Y(y) \int_{- \infty}^{a - y} f_X(x) \mathrm{d}x \, \mathrm{d}y \\
&= \int_{-\infty}^{\infty} F_X(a - y)f_Y(y) \,\mathrm{d}y 
\end{align*}
$$
<p></p>
            حالا تابع توزیع
            $f_{X+Y}$
           را محاسبه می‎کنیم:
           <p></p>
$$
    \begin{align*}
    f_{X+Y} &= \frac{d}{da} \int_{-\infty}^{\infty} F_X(a-y)f_Y(y) dy
    \\ &= \int_{-\infty}^{\infty} \frac{d}{da} F_X(a-y)f_Y(y)\, dy
    \\ &= \int_{-\infty}^{\infty} f_X(a-y)f_Y(y)\, dy \tag{*}
    \\ &= f_X(a)*f_Y(a) 
    \end{align*}
$$
<p></p>
            توزیع حاصل  convolution دو توزیع 
            $f_X(x)$
            و
            $f_Y(y)$
            است.
        </font>
    </div>
</div>
            

<div id="conv1" style="direction:rtl; line-height:300%" align=right>
    <font face="XB Zar" size=6 color=#FF7500>
        توضیحی درباره کانولوشن در حالت پیوسته
    </font>
    <br/>
    <font face="XB Zar" size=5>
        به طور کلی، می‎توان به کانولوشن به شکل‎های متفاوتی نگاه کرد. همچنین شهود مختلفی در کاربردهای بسیار آن در دروس مهندسی وجود دارد. در این قسمت سعی شده توضیحی کلی و شهودی مناسب در ارتباط با درس آمار و احتمالات ارائه شود. فرض کنید دو تابع 
        $ f(t) $
        و
        $ g(t) $
        داریم. 
        تعاریف مختلفی برای کانولوشن وجود دارد اما تعریفی که در این درس کاربرد دارد، عبارتست از:
        $$
        (f * g)(t) = \int_{-\infty}^{\infty} f(t-\tau)g(\tau)d\tau
        $$
        با توجه به اینکه انتگرال روی 
        $ \tau $
        است می‎توان تابع
        $
        f(t - \tau)
        $
        را تابع 
        $
        f(\tau)
        $
        فرض کرد که ابتدا نسبت به محور 
        $ y $
        قرینه شده و بعد 
        $ t $
        واحد به سمت راست شیفت داده شده. (یا اینکه می‎توان آن را 
        $ f(\tau) $
        فرض کرد که ابتدا 
        $ t $
        واحد به سمت چپ شیفت داده شده و بعد نسبت به محور 
        $ y $
        قرینه شده). تابع 
        $ g(\tau) $
        هم بدون تغییر در رابطه قرار دارد. حالا فرض کنید که 
        $$
        f(t) = 
        \left\{
            \begin{array}{ll}
                1 & -\frac{1}{2} \le t \le \frac{1}{2} \\
                0 & \mbox{otherwise}
            \end{array}
        \right.
        $$
        و
        $$
        g(t) =
        \left\{
            \begin{array}{ll}
                \frac{1}{2} & -\frac{1}{4} \le t \le \frac{1}{4} \\
                0 & \mbox{otherwise}
            \end{array}
        \right.
        $$
        اگر بخواهیم مقدار 
        $ (f * g)(t) $
        را در 
        $ t = 0 $
        محاسبه کنیم داریم:
        $$
        (f * g)(0) = \int_{-\infty}^{\infty} f(-\tau)g(\tau)d\tau
        $$
        این رابطه به این معنی است که نمودارهای دو تابع 
        $ f(-\tau) $
        و 
        $ g(\tau) $
        روی هم بگذاریم و مساحت مشترک آن‎ها را حساب کنیم. به این ترتیب مقدار 
        $ (f*g)(0) $
        برابر است با 
        $ \frac{1}{4} $
        . همینطور اگر بخواهیم کانولوشن را نقطه دلخواهی مانند 
        $ t $
        ارزیابی کنیم کافیست مساحت مشترک نمودار تابع 
        $ g(\tau) $
        و نمودار 
        $ f(t - \tau) $
        را محاسبه کنیم. در این مثال به خصوص پاسخ نهایی را می‎توان با توجه به شکل زیر (
        $f$
        قرمز، 
        $g$
         آبی و نتیجه سبز رنگ) محاسبه کرد:
        <p></p>
        <img src="http://mathworld.wolfram.com/images/gifs/convrect.gif" alt="Convolution Wolfram Alpha" title="Convolution Demo" />
        <p></p>
        توجه کنید که در شکل تابعی که شیفت داده می‎شود تابع 
        $g$
        است و نه 
        $f$.
        این در واقع همان خاصیت جابجایی کانولوشن است که در تمرین‎های زیر ثابت می‎کنید!
    </font>
</div>

<div style="direction:rtl; line-height:300%" align=right>
    <font face="XB Zar" size=6 color=#0000FF>
        تمرین
    </font>
    <br/>
    <font face="XB Zar" size=5>
        خاصیت‌های زیر در رابطه با کانولوشن را ثابت کنید
        <ol>
            <li>
            خاصیت جابجایی:
            $$f*g = g*f$$
            </li>
            <li>
                خاصیت پخشی:
                $$
                f*(g+h) = (f*g) + (f*h)
                $$
            </li>
            <li>
                خاصیت شرکت‌پذیری
                $$
                f*(g*h) = (f*g)*h
                $$
            </li>
        </ol>
    </font>
</div>

<div id="two_uniforms" style="direction:rtl; line-height:300%">
    <div align=right>
        <font face="XB Zar" size=6 color=#FF7500>
            مثالی از دو متغیر تصادفی مستقل با توزیع یکنواخت
        </font>
        <p></p>
        <hr/>
        <font face="XB Zar" size=5>
            به طور مثال فرض کنید که دو متغیر تصادفی مستقل با توزیع یکنواخت بین صفر و یک داشته باشیم.
            حالا می‎خواهیم توزیع مجموع این دو متغیر تصادفی را حساب کنیم. روش اول استفاده از روش شهودی کانولوشن است. به این ترتیب که نمودار توزیع یک متغیر را نسبت به محور y قرینه کنیم و بعد نمودار آن را از روی نمودار توزیع متغیر دیگر عبور دهیم.
            اما اینجا می‎خواهیم از فرمول‌های بدست آمده استفاده کنیم.
            طبق توزیع یکنواخت داریم:
            
            
$$
f_X(a) = f_Y(a) =
\left\{
	\begin{array}{ll}
		1  & 0 \lt a \lt 1 \\
		0 & \mbox{otherwise}
	\end{array}
\right.
$$
<p></p>
            با جایگذاری این توزیع‎ها در رابطه کانولوشن داریم:     
$$
f_{X+Y}(a) = \int_{0}^{1} f_X(a-y)\,\mathrm{d}y
$$
<p></p>
            در حالتِ
$$
0 \le a \le 1
$$
            داریم:
$$
f_{X+Y}(a) = \int_0^{a} \mathrm{d}y
$$
<p></p>
            در حالتِ
$$
1 < a \le 2
$$
            داریم:
$$
f_{X+Y}(a) = \int_{0}^{1} f_X(a-y) \, \mathrm{d}y = \int_{a - 1}^{1} \mathrm{d}y
$$
<p></p>
         در نتیجه توزیع نهایی عبارتست از:            
$$
f_{X+Y} = \left\{
            \begin{array}{ll}
            a & 0 \le a \le 1 \\
            2 - a & 1 \lt a \le 2 \\
            0 & \mbox{otherwise}
            \end{array}
            \right.
$$
<p></p>
            با اجرای کد پایین می‎توان درستی نتیجه گیری را مشاهده کرد. به دلیل شکل خاص این توزیع به آن توزیع مثلثی می‎گویند. 
        </font>
    </div>
</div>

```r
library(ggplot2)
n <- 10000
x <- runif(n,0,1)
y <- runif(n,0,1)
z <- x+y
df <- data.frame(x=z)
p1 <- ggplot(data=df, aes(x=x)) + geom_density()
p1
```

<div id="n_rvs" style="direction:rtl; line-height:300%">
    <div align=right>
        <font face="XB Zar" size=6 color=#FF7500>
            مجموع چند متغیر تصادفی مستقل
        </font>
        <p></p>
        <hr/>
        <font face="XB Zar" size=5>
            در این قسمت می‎خواهیم مجموع چند متغیر تصادفی را بررسی کنیم. فرض کنید به جای دو متغیر تصادفی n متغیر تصادفی با توزیع یکنواخت داشتیم. توزیع تجمعی مجموع این n متغیر تصادفی را به شکل زیر تعریف می‎کنیم:
            
$$
F_n(x) = P\{X_1 + X_2 + ... + X_n \le x \}
$$
<p></p>           
با وجود اینکه بدست آوردن یک رابطه صریح در حالت کلی برای 
$ F_n(x) $
دشوار به نظر می‎رسد، اما زمانی که 
$ x \le 1 $
رابطه‎ای بدست می‎آید که ظاهر مناسبی دارد. به طور دقیق‎تر
            
$$
F_n(x) = \frac{x^n}{n!}\,\,\, (0 \le x \le 1)
$$
<p></p>
            در ادامه با استقرا رابطه بالا را ثابت می‎کنیم. 
<p></p>
            پایه استقرا برای $n = 1$ واضح است. در گام استقرا داریم:

$$
F_{n-1}(x) = \frac{x^{n-1}}{(n-1)!} \,\,\, (0 \le x \le 1)
$$
<p></p>
            از طرفی،

$$
\sum\limits_{i=1}^{n} X_i = \sum\limits_{i=1}^{n-1} + X_n \\
$$
<p></p>
            حالا از رابطه‎ای که برای $F_{X+Y}$ در قسمت قبل بدست آمد استفاده می‎کنیم:

$$
\begin{align*}
F_n(x) &= \int_{0}^{1} F_{n-1}(x - y)f_{X_n}(y) \, dy \\
&= \frac{1}{(n-1)!} \int_{0}^{1} (x-y)^{n-1}\, dy \\
\end{align*}
$$
<p></p>
            با توجه به اینکه تمامی $X_i$ها مثبت هستند و همچنین
            $ 0 \le x \le 1$
            می‎توانیم بنویسیم:

$$
\begin{align*}
F_n(x) &= \int_{0}^{x} (x-y)^{n-1}\,dy \\
&= \frac{x^n}{n!}
\end{align*}
$$
<p></p>
            و به این ترتیب اثبات کامل می‎شود.
        </font>
    </div>
</div>

<div id="n_normals" style="direction:rtl; line-height:300%">
    <div align=right>
        <font face="XB Zar" size=6 color=#FF7500>
            مجموع چند متغیرِ تصادفی مستقل با توزیع نرمال
        </font>
        <p></p>
        <hr/>
        <font face="XB Zar" size=5>
            در این قسمت می‎خواهیم قضیه مهمی را در مورد مجموع چند متغیر تصادفی با توزیع نرمال ثابت کنیم.
        </font>
        <br/>
        <font face="XB Zar" size=5 color=#FF0000>
            قضیه: 
        </font>
        <font face="XB Zar" size=5>
            فرض کنید 
            $ X_i, i = 1, 2, 3, ... , n $
            متغیرهای تصادفی مستقل با توزیع نرمال باشند. همچنین فرض کنید پارامترهای این متغیرهای تصادفی به ترتیب 
            $ \mu_i, \sigma_i^2, i = 1, 2, 3, ..., n $
            باشند، در آن صورت متغیر تصادفیِ 
            $ \sum\limits_{i = 1}^{n} X_i $
            از توزیع نرمال با پارامترهای
            $ \sum\limits_{i = 1}^{n} \mu_i$
            و
            $ \sum\limits_{i = 1}^{n} \sigma_i^2$
            است.
<p></p>
برای شروع اثبات، فرض کنید 
$X$
و
$Y$
دو متغیر تصادفی مستقل با توزیع نرمال باشند. همچنین فرض کنید که میانگین
$X$
برابر 0 و واریانس آن برابرِ
$\sigma^2$
باشد. همینطور برای 
$Y$
فرض کنید که میانگین آن 0 و واریانس آن 1 باشد. حالا می‎خواهیم با استفاده از رابطه
$(*)$
تابع توزیع
$X + Y$
را پیدا کنیم.
<br/>
اگر قرار دهیم:
$$
a = \frac{1}{2\sigma^2} + \frac{1}{2}= \frac{1 + \sigma^2}{2\sigma^2}
$$
در آن صورت داریم:
$$
\begin{align*}
f_X(a - y)f_Y(y) &= \frac{1}{\sqrt{2\pi}\sigma}\, \mathrm{exp}\{-\frac{(a-y)^2}{2\sigma^2}\}\frac{1}{\sqrt{2\pi}}\, \mathrm{exp}\{-\frac{y^2}{2}\} \\
&= \frac{1}{2\pi\sigma} \, \mathrm{exp}\{-\frac{a^2}{2\sigma^2} \} \, \mathrm{exp} \{ -c(y^2 - 2y\frac{a}{1+\sigma^2}) \}
\end{align*}
$$
<p></p>
بنابراین با استفاده از رابطه 
$(*)$
می‎توانیم نتیجه بگیریم:
$$
\begin{align*}
f_{X+Y}(a) &= \frac{1}{2\pi\sigma} \, \mathrm{exp}\{\!- \frac{a^2}{2\sigma^2}\! \} \, \mathrm{exp}\{\!\frac{a^2}{2\sigma^2(1 + \sigma^2)} \! \} \int_{-\infty}^{\infty} \mathrm{exp} \{ -c(y - \frac{a}{1 + \sigma^2})^2 \} dy \\
&= \frac{1}{2\pi\sigma} \, \mathrm{exp}\{ -\frac{a^2}{2(a + \sigma^2)} \} \, \int_{- \infty}^{\infty} \mathrm{exp}\{ -cx^2 \} dx \\
&= C \mathrm{exp}\{ -\frac{a^2}{2(a + \sigma^2)} \}
\end{align*}
$$
<p></p>
که اینجا
$ C $
یک ثابت است که به 
$a$
وابسته نیست. این رابطه در حقیقت نشان می‎دهد که متغیرِ تصادفی 
$X + Y$
از توزیع نرمال با میانگین 0 و واریانس 
$ 1 + \sigma^2 $
است.
<p></p>
حالا فرض کنید 
$X_1$
و
$X_2$
دو متغیرتصادفی مستقل با پارامترهای
$
\mu_i
$
و
$
\sigma_i^2
$
باشند. می‎توانیم بنویسیم:
$$
X_1 + X_2 = \sigma_2(\frac{X_1 - \mu_1}{\sigma_2} + \frac{X_2 - \mu_2}{\sigma_2}) + \mu_1 + \mu_2
$$
<p></p>
از آن‎جا که 
$
\frac{X_1 - \mu_1}{\sigma_2}
$
را می‎توان یک متغیر تصادفی نرمال با میانگین 0 و واریانسِ 
$
\frac{\sigma_1^2}{\sigma_2^2}
$
در نظر گرفت و همچینین 
$
\frac{X_2 - \mu_2}{\sigma_2}
$
را یک متغیر تصادفی نرمال با میانگین 0 و واریانس 1، طبق نتیجه‎ای که در ابتدای اثبات گرفتیم، نتیجه می‎شود که 
$$
\frac{X_1 - \mu_1}{\sigma_2} + \frac{X_2 - \mu_2}{\sigma_2}
$$
<p></p>
یک متغیر تصادفی نرمال با میانگین 0 و واریانسِ 
$
1 + \frac{\sigma_1^2}{\sigma_2^2}
$
است. بنابراین 
$
X_1 + X_2
$
یک متغیر تصادفی با میانگینِ 
$
\mu_1 + \mu_2
$
و واریانسِ 
$
\sigma_2^2 (1 + \frac{\sigma_1^2}{\sigma_2^2}) = \sigma_1^2 + \sigma_2^2
$
است.
<br/>
بنابراین قضیه برای 
$
n = 2
$
ثابت شد. برای حالات بیشتر n می‎توان اثبات را با استقرا ادامه داد که این بخش به عنوان تمرین در ادامه قرار داده شده است. اما برای شهود بهتر، جمع دو متغیر تصادفی مستقل با توزیع نرمال در نمودار زیر نشان داده شده است.
        </font>
    </div>
</div>

<div id="discrete" style="direction:rtl;line-height:300%">
    <div align=right>
        <font face="XB Zar" size=6 color=#FF7500>
            بررسی جمع دو متغیر تصادفی مستقل در حالت گسسته
        </font>
        <p></p>
        <hr/>
        <font face="XB Zar" size="5">
            در این قسمت می‎خواهیم جمع دو متغیر تصادفی مستقل و گسسته را بررسی کنیم. سعی می‎کنیم حالت گسسته را با یک مثال توضیح دهیم.
            <br/>
            فرض کنید دو متغیر تصادفی
            $
            X
            $
            و
            $
            Y
            $
            با توزیع پواسون و به ترتیب با پارامترهای
            $
            \lambda_1
            $
            و
            $
            \lambda_2
            $
            داریم. می‎خواهیم توزیع 
            $
            X + Y
            $
            را محاسبه کنیم. 
            <br/>
            با توجه به اینکه در حالت گسسته هستیم باید 
            $
            P\{X + Y = n \}
            $
            را محاسبه کنیم. این احتمال را اینگونه حساب می‎کنیم:
            $$
            \begin{align*}
            P\{X + Y = n\} &= \sum\limits_{k = 0}^{n} P\{X = k , Y = n - k\} \\
            &= \sum\limits_{k = 0}^{n} P\{X = k\}P\{Y = n - k\} \\
            &= \sum\limits_{k = 0}^{n} e^{-\lambda_1}\frac{\lambda_1^k}{\lambda_1!} \, e^{-\lambda_2}\frac{\lambda_2^{n-k}}{(n-k)!} \\
            &= e^{-(\lambda_1 + \lambda_2)} \sum\limits_{k = 0}^{n} \frac{\lambda_1^k \lambda_2^{n-k}}{k! (n-k)!} \\
            &= \frac{e^{-(\lambda_1 + \lambda_2)}}{n!} \sum\limits_{k = 0}^{n} \frac{n!}{k! (n-k)!} \lambda_1^k \lambda_2^{n-k} \\
            &= \frac{e^{-(\lambda_1 + \lambda_2)}}{n!} (\lambda_1 + \lambda_2)^n
            \end{align*}
            $$
            <p></p>
            بنابراین،
            $
            X + Y
            $
            از یک توزیع پواسون با پارامتر 
            $
            \lambda_1 + \lambda_2
            $
            می‎آید. سعی شده در هیستوگرام زیر این نتیجه نشان داده شود:
        </font>
    </div>
</div>
