احتمال شرطی
        </font>
		<p></p>
		<hr>
احتمال شرطی یکی از مباحث مهم و اصلی در حوزه ی احتمال است. احتمال رخداد واقعه $A$ به شرط رخداد واقعه $B$ را به صورت زیر تعریف می‌کنیم:
        $$P(A|B) = \frac{P(A \cap B)}{P(B)}$$
        با این فرض که $P(B) > 0$
        <br>
         <figure>
  <img src="figs/conditional_b.png" style="float:center; width:40%"/>
  <figcaption>
    <div align="center">
                فضای احتمال $A|B$
    برابر است با فضایی از $A$ که در 
   $B$
    دیده می‌شود.
    </div>
    </figcaption>
</figure> 
        <br/>
        می‌دانیم احتمال وقوع هر پدیدهٔ تصادفی (پیشامد) برابر است با نسبت تعداد اعضای آن پدیده (پیشامد) به تعداد اعضای فضای نمونه. در احتمال شرطی، احتمال وقوع پیشامد، $P(A\cap B)$ است که بیانگر احتمال وقوع همزمان پیشامدهای A و B می‌باشد، و با توجه به اینکه می‌دانیم B حتماً اتفاق افتاده، فضای نمونه به B کاهش می‌یابد و نسبت مذکور به صورت فوق محاسبه خواهد شد.  
        <br>
        در واقع تقسیم بر P(B) به منظور نرمال سازی انجام می شود تا مجموع احتمال پیشامد ها یک بماند.
        <br>
        <br>
        برای مثال،اگر واقعه $A$ احتمالی برابر با ۰.۳، واقعه $B$ احتمالی برابر با ۰.۲ و واقعه $A \cap B$ احتمالی برابر با ۰.۰۱ داشته باشد، احتمال $A|B$ برابر است با:
        $$P(A|B) = \frac{P(A \cap B)}{P(B)} = \frac{0.01}{0.2} = \frac{1}{20}$$
        <br>
        <br/>
        حال به حل مثالی دیگر می پردازیم. فرض کنید تاسی را پرتاب کرده‌ایم. با این فرض که می دانیم عدد به دست آمده عددی زوج است، احتمال این که عدد به دست آمده از ۴ کمتر باشد چقدر است؟
        <br>
        <br/>
        پاسخ: فرض کنید واقعه $A$ را برابر با رخداد عدد کمتر از ۴ در نظر بگیریم و واقعه $B$ را برابر با رخداد عددی زوج بگیریم. حال خواهیم داشت:
        $$P(A|B) = \frac{P(A \cap B)}{P(B)}$$
        $$P(A \cap B) = \frac{1}{6}$$
        $$P(B) = \frac{1}{2}$$
        $$ \Longrightarrow P(A|B) = \frac{\frac{1}{6}}{\frac{1}{2}} = \frac{1}{3}$$
        <br>
        <br/>

قانون احتمال کل
        </font>
		<p></p>
		<hr>
قانون احتمال کل ارتباط میان یک پیشامد و افرازهای فضای نمونه به بخش های مختلف را بیان می کند. فرض کنید $B_1,B_2,...,B_m$ افراز فضای نمونه $\Omega$ به $m$ زیربخش باشد(یعنی $B_i$ ها دو به دو اشتراکی ندارند و اجتماع آن ها کل فضای $\Omega$ را تشکیل می دهد). حال اگر $A$ را پیشامد رخ داده در $\Omega$ درنظر بگیریم، داریم: 
        $$P(A) = P(A|B_1)P(B_1) + P(A|B_2)P(B_2) + ... + P(A|B_m)P(B_m)$$
        از این رابطه زمانی استفاده می شود که محاسبه احتمال شرطی پیشامد $A$ از محاسبه مستقیم این پیشامد راحت تر است.
        <br>
        <br/>
        <font color=#00a339 size=5>
مثال:
        </font>
 فرض کنید دو جعبه لامپ از کارخانه رسیده است. جعبه اول شامل ۲۱ لامپ است که ۱۲ تای آن سالم و ۹ تای آن خراب است. جعبه دوم شامل ۲۱ لامپ است که ۱۰ تای آن سالم و ۱۱ تای آن خراب است. لامپی را از جعبه بیرون می‌آوریم. احتمال این که آن لامپ خراب باشد چقدر است؟
        <br>
        <br/>
        <font color=#00a339 size=5>
        پاسخ:
        </font>
        داریم:
        $$P(Bad \ light \ bulb) = P(Bad \ light \ bulb \ | \ Box \ 1)P(Box \ 1) + P(Bad \ light \ bulb \ | \ Box \ 2)P(Box \ 2)$$
        $$ \Longrightarrow P(Bad \ light \ bulb) = \frac{1}{2} \cdot \frac{9}{21} + \frac{1}{2} \cdot \frac{11}{21} = \frac{10}{21}$$
	</font>
</div>

<div id="sec_bayes" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
قاعده بیز
        </font>
		<p></p>
		<hr>
در تمامی بحث‌های پیشین، هدف ما یافتن احتمال رخداد یک پیشامد $A$ به شرط وقوع پیشامد $B$ بود. اما گاهی این موضوع پیش می‌آید که ما احتمال رخداد واقعه $B$ به شرط وقوع واقعه $A$ را نیاز داریم. برای محاسبه این احتمال، باید از قاعدهٔ بیز کمک بگیریم. قاعده بیز به صورت زیر بیان می‌شود:
        $$P(B|A) = \frac{P(A|B)P(B)}{P(A)}$$
        درستی رابطه بالا به صورت زیر اثبات می شود:
        <br>
        $$P(A|B) = \frac{P(A \cap B)}{P(B)} \Longrightarrow P(A \cap B) = {P(A|B)}{P(B)}$$
        $$\Longrightarrow P(B|A) = \frac{P(B \cap A)}{P(A)} = \frac{(P(A \cap B)}{P(A)} = \frac{P(A|B)P(B)}{P(A)}$$
        در حالت کلی قاعده بیز به صورت زیر بیان می شود
        <br>
        فرض کنید $B_1,B_2,...,B_m$ افراز فضای نمونه $\Omega$ به $m$ زیربخش باشد در این صورت برای هر i داریم:
        $$P(B_i|A) = \frac{P(B_i) \, P(A|B_i)}{\sum_{j=1}^m P(B_j) \,P(A|B_j)}$$ 
        <br>
        برای اثبات کافی است توجه کنیم که صورت کسر سمت راست طبق تعریف احتمال شرطی برابر با $P(B_i \cap A)$ است و مخرج کسر طبق قانون احتمال کل برابر با $P(A)$
        <br/>
        <font color=#00a339 size=5>
مثال:
        </font>
 در مثال قسمت قبل، احتمال این که لامپ خراب متعلق به جعبه دوم باشد چقدر است؟
        <br>
        <br/>
        <font color=#00a339 size=5>
        پاسخ:
        </font>
        داریم:
        $$P(Box \ 2 \ | \ Bad \ light \ bulb) = \frac{P(Bad \ light \ bulb \ | \ Box \ 2)P(Box \ 2)}{P(Bad \ light \ bulb)}$$
        $$ \Longrightarrow P(Box \ 2 \ | \ Bad \ light \ bulb) = \frac{\frac{1}{2} \cdot \frac{11}{21}}{\frac{10}{21}} = \frac{11}{20}$$
	</font>
</div>

<div id="sec_problem" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
بازگشت به مسأله مسابقه تلویزیونی
        </font>
		<p></p>
		<hr>
واقعه های $A$، $B$ و $C$ را به ترتیب وجود داشتن ماشین در درهای ۱، ۲ و ۳ بگیرید. به وضوح می دانیم:
        $$P(A) = P(B) = P(C) = \frac{1}{3}$$
        حال بدون از دست دادن کلیت فرض کنید ما در شماره ۱ را انتخاب کرده باشیم و مجری در ۲ را برایمان باز کند.
         احتمال رخداد $B$ را توسط مجری با $P(HB)$ نشان می دهیم. حال روابط زیر به وضوح برقرارند:
        $$P(HB|A) = \frac{1}{2}, \ P(HB|B) = 0, \ P(HB|C) = 1$$
        برای بررسی تعویض در، کافیست احتمال های $P(A|HB)$ و $P(C|HB)$ را محاسبه کنیم. طبق قاعده بیز داریم:
        $$P(A|HB) = \frac{P(HB|A)P(A)}{P(HB)} = \frac{P(HB|A)P(A)}{P(HB|A)P(A) + P(HB|B)P(B) + P(HB|C)P(C)} = \frac{\frac{1}{2} \cdot \frac{1}{3}}{\frac{1}{2} \cdot \frac{1}{3} + 0 \cdot \frac{1}{3} + 1 \cdot \frac{1}{3}} = \frac{1}{3}$$
        $$P(C|HB) = \frac{P(HB|C)P(C)}{P(HB)} = \frac{P(HB|C)P(C)}{P(HB|A)P(A) + P(HB|B)P(B) + P(HB|C)P(C)} = \frac{1 \cdot \frac{1}{3}}{\frac{1}{2} \cdot \frac{1}{3} + 0 \cdot \frac{1}{3} + 1 \cdot \frac{1}{3}} = \frac{2}{3}$$
        بنابراین، تعویض در انتخاب کار معقولانه‌ای خواهد بود.
        <br>
       برای درک بهتر میتوانید از کد زیر استفاده کنید و جواب را با جواب به دست آمده مقایسه کنید.
	</font>
</div>

```r
monty <- c()
doors <- c("A", "B", "C")
for (i in 1:1000){
  prize <- sample(doors, 1)
  pick <- sample(doors, 1)
  open <- sample(doors[doors != pick & doors != prize], 1)
  to_switch <- doors[doors != pick & doors != open]
  if (pick == prize){
    monty <- c(monty, 'noSwitch')
  }
  if (to_switch == prize){
    monty <- c(monty, "Switch")
  }
}
cat("Probability to win if you don't switch: ", length(monty[monty == 'noSwitch'])/ (length(monty[monty == 'Switch']) + length(monty[monty == 'noSwitch'])),"\n")
cat("Probability to win if you switch: ", (length(monty[monty == 'Switch'])) / (length(monty[monty == 'Switch']) + length(monty[monty == 'noSwitch'])))
```

مسئله ای برای جمع بندی
        </font>
    <br/>
    <br/>
        <font color=#00a339 size=5>
مثال:
        </font>
     فرض کنید در کیفی ، یک توپ وجود دارد که رنگ آن یا آبی است یا سبز(با احتمالی مساوی).یک توپ سبز در کیف قرار می دهیم(یعنی الان دو توپ در کیف قرار دارد) و بعد به صورت تصادفی یکی از توپ ها را بیرون می آوریم و می بینیم که توپ بیرون آمده به رنگ سبز است.احتمال اینکه توپ باقی مانده در کیف نیز سبز باشد چقدر است؟
        <br>
        <br/>
        <font color=#00a339 size=5>
        پاسخ:
        </font>
        $A$ را پیشامد سبز بودن توپ اولیه در کیف در نظر می گیریم.$B$ را پیشامد اینکه توپ بیرون آورده شده سبز باشد در نظر بگیرید و همچنین $C$ را پیشامد اینکه توپ باقی مانده در کیف سبز باشد در نظر بگیرید.
        <br/>
        حالا ما برای حل مسئله نیاز به پیدا کردن $P(C|B)$ داریم
        $$P (C|B) = P (C|B, A)P (A|B) + P (C|B, A^{C})P (A^{C}|B) = 1P (A|B) + 0P (A^{C}|B)$$
        برای پیدا کردن $P(A|B)$ از قاعده بیز استفاده می کنیم.
        $$P(A|B) = \frac{P(B|A)P(A)}{P(B)} = \frac{1/2}{P(B|A)P(A) + P(B|A^{c})P(A^{c})} = \frac{1/2}{1/2 + 1/4} = \frac{2}{3}$$
	</font>
</div>

خلاصه و نتیجه‌گیری
        </font>

		<hr>
        در این بخش با موارد زیر آشنا شدیم:
        <ul>
            <li>
            احتمال شرطی
            </li>
            $$P(A|B) = \frac{P(A \cap B)}{P(B)}$$
            <li>
            به کار گیری احتمال شرطی و استقلال برای حل مسائل احتمال
            </li>
            به وابسته‌ها و وابستگی‌ها اهمیت بده!
            <li>
            قانون احتمال کل
            </li>
            $$P(A) = P(A|B_1)P(B_1) + P(A|B_2)P(B_2) + ... + P(A|B_m)P(B_m)$$
            <li>
            قاعده بیز
            </li>
            $$P(B|A) = \frac{P(A|B)P(B)}{P(A)}$$
            <li>
            مسأله مسابقه تلویزیونی
            </li>
