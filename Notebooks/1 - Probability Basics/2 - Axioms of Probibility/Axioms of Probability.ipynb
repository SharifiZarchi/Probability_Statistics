مقدمه
        </font>
		<p></p>
		<hr>
در نظریه احتمال سر و کله ی مجموعه ها پیدا می شود. در این قسمت ابتدا می خواهیم ردپای نظریه مجموعه ها را در احتمالات دنبال کنیم.
سپس اجزای مختلف فضای احتمال را معرفی می کنیم و مفهوم احتمال برای مجموعه های مختلف را مطالعه خواهیم کرد.
<br/>
یکی از راه ها برای فکر کردن درباره امر تصافی بودن این است که سعی کنیم راهی پیدا کنیم تا چگونه چیزی را که نمی دانیم ابراز کنیم. وقتی می گوییم چیزی تصادفی است می خواهیم بگوییم که دانش ما در مورد پیشامد محدود است، پس با قطعیت نمی توان گفت چه اتفاقی خواهد افتاد.
<br/><br/>
	</font>
</div>

<div id="pr_ex" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مثالی از کابرد احتمالات
        </font>
		<p></p>
		<hr>
        مخابرات در زندگی روزمره ما نقش ویژه ای ایفا می کند. برای مثال هنگامی که با تلفن همراه خود صحبت می کنید، هنگامی که به شبکه اینترنت متصل می شوید و یا وقتی که از یک کنترل از راه دور برای کنترل استفاده می کنید از مخابرات کمک می گیرید. همه این کارها بر انتقال اطلاعات استوار هستند. برای مثال وقتی شما با تلفن همراه خود صحبت می کنید صدای شما به رشته ای از بیت ها تبدیل می شود که به آن ها information bits گفته می شود.
        <br/>
        مشکلی در ارسال بیت ها وجود دارد آن هم این است که نویز این روند را تحت تاثیر قرار می دهد و ممکن است برای مثال رشته "...1010" به رشته "...1011" تبدیل شود که در این جا بیت چهارم دچار خطا شده است. این خطا ها کیفیت مخابرات را تحت تاثیر قرار می دهند و مثلا می تواند باعث شود صدای شما کیفیت خود را از دست بدهد.
        <br/>
        نویز یک پدیده تصادفی است که ما قبل از ارسال بیت ها نمی دانیم کدام بیت ممکن است دچار خطا شود. از تئوری احتمالات در مخابرات به شدت استفاده می شود برای مثال می توان برای مدل کردن رفتار نویز به احتمالات رجوع کرد و درنهایت حتی می توان سیستمی در نظر گرفت تا بیت هایی که دچار خطا می شوند را تصحیح کرد.
<br/><br/>
	</font>
</div>

<div id = "prob_ex" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=blue size=6>
پیدایش احتمال 
        </font><br/>
        ریاضیدان ها به دنبال ارائه ای از اصولی بودند تا بتوانند مسایل احتمالی را توصیف و فرمول بندی کنند. مثل اصول هندسه اقلیدس یا اصول مکانیک کلاسیک نیوتون.
        <br/>
        یک تعریف پیشنهادی برای احتمال پیشامد $A$ که به ذهن می رسد نسبت تعداد دفعاتی است که پیشامد $A$ در $n$ آزمایش رخ می دهد.
        <br/>
        برای مثال اگر یک سکه را 1000 دفعه پرتاب کنیم و 800 مرتبه سکه شیر بیاید و 200 بار خط بیاید، می توانیم بگوییم احتمال شیر آمدن در پرتاب سکه برابر 0.8 است، یا به عبارتی اگر این سکه را 10 بار دیگر پرتاب کنیم انتظار داریم 8 مرتبه به شیر برسیم و 2 بار خط را ببینیم. البته به طور حتم این استدلال دقیق نیست. مثلا ممکن است سکه را 10000 مرتبه پرتاب کنبم و ببینیم 7500 بار شیر آمد و 2500 مرتبه خط. پس نیاز داریم استدلال خود را بسته به دقتی که به دنبال آن هستیم تصحیح کنیم.
        <br/>
        به نظر می رسد هر چه $n$ را بزرگ تر کنیم به نتیجه دقیق تری خواهیم رسید. بر این اساس عده ای تعریف زیر را برای احتمال در نظر گرفتند:
        <br/>
        $Pr(A) = \lim_{n \to \infty} \frac{n(A)}{n}$
        <br/>
        اما این تعریف چند اشکال داشت: 
        <ul>
        <li>
        این حد ممکن است وجود نداشته باشد.
        </li>
        <li>
        شاید این حد مقدار ثابتی را ندهد، یعنی در آزمایش های مختلف جواب های گوناگونی بگیریم.
        </li>
        <li>
        این که $n$ را به سمت بی نهایت میل دهیم در دنیای واقعی امکان ناپذیر است.
        </li>
        </ul>
        امروزه ریاضی دان ها از اصول کولموگروف برای برخورد با احتمالات بهره می گیرند که در ادامه آن ها را مطرح می کنیم.
        <br>
        <br/>
        <li>
        سوال: $n$ نفر در یک مهمانی حضور دارند.احتمال این که هیچ دو تایی از این افراد در یک روز از سال به دنیا نیامده باشند را حساب کنید.برای این که این احتمال از 0.5 کمتر باشد $n$  باید چه شرطی داشته باشد؟
        </li>
	</font>
</div>

		<font color=#FF7500 size=6>
یادآوری نظریه مجموعه ها
        </font>
		<p></p>
        <hr>
        <ul>
        <li>
       مجموعه (Set) : مجموعه دسته ای از اشیا می باشد و هر شی عضوی از مجموعه می باشد. برای عضو $x$ که در مجموعه $S$ قرار دارد می گوییم $x$ عضو $S$ است با $x ∈ S$ نمایش می دهیم.
       </li>
       </ul>
       <ul>
       <li>
        عضویت: اگر شی $x$ در میان اشیایی که در $S$ هستند نباشد آنگاه $x$ عضو $S$ نیست و می نویسیم $x \notin S$.
        </li>
        </ul>
        <ul>
        <li>
        تعداد اعضای مجموعه می تواند محدود، نامحدود و شمارا(متناظر با اعداد طبیعی) و یا غیر قابل شمارش باشد. ترتیب اعضا در مجموعه نیز اهمیتی ندارد.
        </li>
        </ul>
        <ul>
        <li>
        مجموعه $B$ را زیر مجموعه مجموعه $A$ می گوییم اگر و تنها اگر هر عضو از مجموعه $B$ عضوی از مجموعه $A$ باشد می نویسیم $A \subset B$.
        در غیر این صورت $B$ زیر مجموعه $A$ نیست و می نویسیم: $B \not\subset A$.
        </li>
        </ul>
        <ul>
        <li>
        مجموعه $A$ را مساوی مجموعه $B$ می گوییم اگر و تنها اگر $A \subset B$ و $B \subset A$.
        </li>
        </ul>
        <ul>
        <li>
        مجموعه‌ای که شامل همه عناصر مورد نظر است را مجموعه مرجع $\Omega$ می گوییم.
        </li>
        </ul>
        <ul>
        <li>
        مجموعه ای که عضوی ندارد را تهی می نامیم و با $\varnothing$ نمایش می دهیم.
        </li>
        </ul>
        <ul>
        <li>
       اجتماع دو مجموعه $A$ و $B$ مجموعه تمام عناصری است که در $A$ یا در $B$ یا در هر دوی آن ها باشند و اجتماع دو مجموعه را با $A \cup B$ نمایش می دهیم.
       </li>
       </ul>
       <img src="Venn_A_union_B.png" style="float:center; width:50%"/>
       <ul>
       <li>
           اشتراک دو مجموعه (مجموعه‌ی
           $A \cap B$) مجموعه عناصری است که هم در $A$ و هم در $B$ هستند.
       </li>
       </ul>
       <img src="A_int_B.png" style="float:center; width:50%"/>
       <ul>
       <li>
       دو مجموعه را جدا از هم (Disjoint) می گوییم اگر و تنها اگر $A \cap B = \varnothing$ باشد.
       </li>
       </ul>
       <img src="dis.png" style="float:center; width:50%"/>
	</font>

        $(A \cup B)\cap C = (A \cap C)\cup(B \cap C)$

        <br/>

        <br/>

نمودار ون 
        </font><br/>
        نمودار وِن یا نمودار مجموعه تصاویری است که در شاخه‌ای از ریاضیات به نام نظریه مجموعه‌ها به کار می‌رود که در سال ۱۸۸۱ توسط جان ون اختراع شد. این نمودار تمام روابط ممکن ریاضی یا منطقی مابین مجموعه‌ها را نشان می‌دهد و معمولاً از دایره‌های همپوشان تشکیل شده است.  نمودار ون به نسبت تعداد دوایر موجود در آن به چند بخش تقسیم میشود. برای مثال در دایره های دو تایی به 4 بخش: دایره A، دایره B، اشتراک و اجتماع آن که جزیی از A و B است و در انتها بنا به تعریف مجموعه مرجع قرار دارد. A-B ( تعداد اعضای مجموعه) برابر است با : تعداد اعضای مجموعه اول - تعداد اعضای مشترک با B و همین قضیه برای اعضای B نیز صادق است، البته این اعداد فقط تعداد اعضایی که در A هست و در B نیست را توضیح می دهد.

فضای احتمال و واقعه:
        </font>
		<p></p>
		<hr>
        فضای احتمال یا مدل احتمالاتی در یک آزمایش از عوامل زیر تشکیل می شود:
        <ul>
            <li>
           مجموعه $\Omega$ شامل کلیه نتایج ممکن برای آزمایش.
            </li>
            <li>
            زیر مجموعه های $\Omega$ که واقعه نامیده می شوند.
            </li>
            <li>
           عدد $P(A)$ که به هر یک از پیشامد واقعه ها نسبت داده می شود.(طبق اصول موضوعه)
            </li>
        </ul>
             فضای نمونه $\Omega = (Sample Space)$ مجموعه کلیه نتایج ممکن برای یک آزمایش تصادفی است:   
        <ul>
            <li>
          برای مثال در آزمایش انداختن سکه داریم:      $\Omega = \{H , T\}$ که داری 2 عضو می باشد.
            </li>
            <li>
            برای مثالی دیگر در آزمایش پرتاب دو سکه داریم:        $\Omega = \{(H,H) , (H,T) , (T,H) , (T,T)\}$ که دارای 4 عضو می باشد.
            </li>
            <li>
            اگر آزمایش پرتاب 2 تاس باشد، فضای نمونه شامل 36 نقطه است.
            </li>   
            <li>
            اگر آزمایش طول عمر(به ساعت) یک ترانزیستور باشد آنگاه فضای نمونه شامل تمام اعداد حقیقی نامنفی است:
            $S = \{x : 0 \leqslant x < \infty\}$
 </li>
        </ul>
        واقعه(Event): هر زیر مجموعه از فضای نمونه‌ها را واقعه یا پیشامد می گویند.
        <ul>
            <li>
            در آزمایش انداختن دو سکه داریم: $A = \{(H,H) , (H,T) , (T,H)\}$
            که $A$ واقعه این است که حداقل 1 شیر بیاید.         
            </li>      
        </ul>
       <ul>
            <li>
            برای مثالی دیگر در آزمایش پرتاب دو سکه داریم:        $\Omega = \{(H,H) , (H,T) , (T,H) , (T,T)\}$ که دارای 4 عضو می باشد.
            </li>        
            <li>
            در مثال طول عمر ترانزیستور $E = \{x : 0 \leqslant x < 5\}$ پیشامد آن است که طول عمر ترانزیستور از 5 بیشتر نباشد.
            </li>
      </ul>
        پیشامد های $E$ و $F$ را جدا از هم یا ناسازگار می گوییم هر گاه مجموعه های $A$ و $B$ جدا از هم باشند:    $A \cap B = \varnothing $     
	</font>
</div>

<div id="Axiomatic" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
تعریف $Axiomatic$:
        </font>
		<p></p>
		<hr>
        اصول کولموگروف: به هر واقعه A عدد $P(A)$ نسبت داده می شود به طوری که:
        <ul>
            <li>
            اصل(1) : $P(A) \geqslant 0$
            </li>
            <li>
            اصل(2) : $P(\Omega) = 1$
            </li>
            <li>
            اصل(3) : اگر واقعه های $A$ و $B$ ناسازگار باشند ($A \cap B = \varnothing$)، آنگاه: $P(A \cup B) = P(A) + P(B)$
        </ul>
        <br/>
        با تکرار اصل 3 می توان آن را برای هر تعداد محدودی از وقایع بیان کرد ولی برای تعداد نامحدود به جای اصل 3 اصل قوی تری را باید جایگزین کرد:
        <br/>
        اصل $(3')$: اگر $A_1 , A_2 , ... $ دو به دو ناسازگار باشند، داریم:
        <br/>
        $P(A_1 \cup A_2 \cup ... ) = P(A_1) + P(A_2) + ... $که این را اصل جمع پذیری نامحدود(قابل شمارش) می گویند.
	</font>
</div>

        گاهی اوقات به آسمان نگاه می کنیم و می گوییم به احتمال $90%$ درصد باران می بارد. احتمال هایی که در چنین جمله ای بدان اشاره می شود اندازه اعتقاد فرد گوینده را بیان می کند. روشن است که این تعریف وابسته به فرد است و نمی تواند مبنای یک تئوری ریاضی و مستحکم قرار گیرد. اما می توان نتیجه گیری های معقولانه هم داشت. به نظر منطقی می رسد که احتمال به اندازه اعتقاد نیز باید اصول احتمال را رعایت کند. مثلا اگر کسی می گوید $70%$ درصد مطمئن هستیم که کتاب "خورشید به دور زمین می گردد" را امین نوشته است و $20%$ درصد مطمئن هستیم که محمد نوشته است، منطقی است بگوییم به احتمال $90%$ درصد یکی از این دو نفر کتاب را نوشته است.
        <br/>
        تعریف فراوانی(فرکانس) نسبی (تعریف آماری یا تجربی):
        <br/>
        اگر یک آزمایش تصادفی را n بار انجام دهیم و در این n بار، $n_A$ بار واقعه A اتفاق افتد، علی الاصول برای n بزرگ، $P(A)$ عددی نزدیک به $\frac{n_A}{n}$ خواهد بود، این دیگر subjective نیست و قابل بررسی تجربی است.
	</font>

وقایع مستقل و وابسته
        </font>
        <hr/>
        <br/>
      در مطالعه احتمالات معمولا با دو دسته از وقایع مواجه می شویم: وقایع مستقل و وقایع وابسته 
        <ul>
            <li>
           وقایع مستقل: به دو پیشامد مستقل می گوییم اگر وقوع یکی بر احتمال وقوع دیگری تاثیر نگذارد. برای مثال در پرتاب یک تاس و سکه، عددی که از تاس به دست می آید ارتباطی با این که سکه خط آمده یا شیر ندارد. به عبارت دیگر دو پیشامد $A$ و $B$ را مستقل می گوییم اگر: $P(A \cap B) = P(A).P(B)$
            </li>
            <li>
            وقایع وابسته: دو پیشامد را وابسته می گوییم اگر احتمال وقوع یکی بر احتمال وقوع دیگری تاثیر گذارد. در زندگی روزمره هم از این مفهوم استفاده می کنیم، مثلا دزدی کردن و به زندان رفتن، تلاش کردن و به موفقیت رسیدن و ... .
            </li>
            <li>
            نکته مهمی که باید به آن دقت شود این است که جدا یا ناسازگار بودن دو پیشامد لزوما به معنای مستقل بودن آن دو پیشامد نیست. برای روشن شدن این موضوع به مثال ساده زیر توجه کنید:
            <br/>
            در یک کیسه 30 توپ داریم. 17 تا از آن ها آبی، 11 تا زرد رنگ و 2 تای باقی مانده سبز هستند. حال یک توپ به تصادف از کیسه خارج می کنیم.
            <br/>
            i) احتمال این که توپ به رنگ آبی باشد برابر $\frac{17}{30}$ است.
            <br/>
            ii) احتمال این که این توپ به رنگ زرد باشد برابر $\frac{11}{30}$ است.
            <br/>
            iii) آیا دو پیشامد بالا ناسازگار هستند؟
            <br/>
            از آن جا که یک توپ نمی تواند هم آبی و هم زرد باشد، به عبارت دیگر استراک این دو پیشامد تهی است پس این دو پیشامد ناسازگار هستند.
            <br/>
            iv) آیا این دو پیشامد مستقل هستند؟ 
            <br/>
            جواب این سوال با توجه به به تعریف استقلال دو پیشامد منفی می باشد.
<br/>
            سوال:
            <br/>
            نشان دهید اگر دو پیشامد A و B مستقل باشند آنگاه $\overline{A}$ و $B$ نیز مستقل اند. 
            <br/>
            نشان دهید اگر دو پیشامد $A$ و $B$ مستقل باشند آنگاه $\overline{A}$ و $\overline{B}$ نیز مستقل اند.
            </li>
            <li>
            سه واقعه A و B و C را مستقل می گوییم هر گاه: 
           <br/>
            $P(A \cap B) = P(A).P(B)$   و
            <br/>
            $P(A \cap C) = P(A).P(C)$   و
            <br/>
            $P(B \cap C) = P(B).P(C)$   و
            <br/>
            $P(A \cap B \cap C) = P(A).P(B).P(C)$
            <br/>
            یعنی اگر سه واقعه مستقل باشند، دو به دو نیز مستقل اند ولی عکس آن لزوما صحیح نیست.
            <br/>
            <br/>
            سوال: نشان دهید اگر A و B و C مستقل باشند آنگاه A از $\overline{BC}$ و $B + C$ مستقل است.
            <br/>
            </li>
            <li>
            n واقعه $A_1$، $A_2$، ... و $A_n$ را مستقل گویند هرگاه برای هر دسته اعداد صحیح $k_1$، $k_2$، ... و $k_r$ ($r \leqslant n$) داشته باشیم:
            <br/>
            $P(A_{k_1} \cap A_{k_2} \cap ... \cap A_{k_r}) = P(A_{k_1})P(A_{k_2})...P(A_{k_r})$
            <br/>
            در واقع تعداد $2^{n} - (n + 1)$ رابطه باید برقرار باشد.
            </li>
        </ul>
</font>
</div>
