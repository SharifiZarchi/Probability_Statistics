مقدمه
        </font>
		<p></p>
		<hr>
             فرض کنید دو سکه‌ی سالم را پرتاب می‌کنیم. آیا احتمال شیر آمدن سکه‌ی دوم به وضعیت سکه‌ی اول وابسته است؟ 
        <br>
        <br/>
         در مثالی دیگر فرض کنید که دو ظرف داریم. در ظرف اول ۲ توپ سفید و ۲ توپ سیاه موجود است. همچنین در ظرف دوم ۳ توپ سفید و ۱ توپ سیاه داریم.حال رویداد B را انتخاب توپ سفید و رویداد A را انتخاب ظرف اول تعریف می‌کنیم. در این شرایط احتمال $P(A \cap B)$ را بدست آورید. 

<p></p>
<br />
<div id="sec_cp" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
استقلال
        </font>
		<p></p>
		<hr>فرض کنید دو رویداد A و B به گونه‌ای تعریف شده باشند که دانستن رخداد یکی , تاثیری در وضعیت دیگری نداشته باشد. در این صورت این دو رویداد را مستقل می‌نامیم. ابتدا دو مثال را بررسی کرده و سپس تعریف علمی آن را بیان می‌کنیم.
        <br>
        <br>
        <font color=#2C5494 size=5>
مثال ۱
        </font>
        <br/>در یک مثال ساده فرض کنید می‌خواهیم احتمال follow back توسط یک فرد در instagram را بدست آوریم. اگر رویداد A را "follow back توسط آن فرد" و رویداد B را "از دماغ فیل افتادن این شخص" تعریف کنیم , پر واضح است که این دو رویداد مستقل نیستند و وابسته‌اند. زیرا اگر بدانیم که فرد مورد نظر از دماغ فیل افتاده , احتمال follow back بسیار کم خواهد بود.  
        <br><br>
        <font color=#2C5494 size=5 id="taas">
مثال ۲
        </font>
        <br/>
        حال به حل مثالی دیگر می پردازیم. فرض کنید  دو تاس را پرتاب کرده‌ایم. رویداد A را "۴ آمدن تاس نخست" تعریف می‌کنیم و  B را رویدادی در نظر می‌گیریم که در آن "جمع دو تاس برابر ۶ شود".    
        حال آیا این دو رویداد مستقل‌اند؟
        <br>
        <br/>
        پاسخ:
        خیر این دو وابسته‌اند . زیرا اگر بدانیم که رویداد B رخ داده به این معنا است که می‌دانیم تاس اول 6 نیامده و این احتمال رویداد A را بیشتر می‌کند.
        <br>
        <br/>
         حال اگر C را رویدادی در نظر بگیریم که در آن "جمع دو تاس برابر ۷ شود".    
        حال آیا رویداد A و C مستقل‌اند؟ 
        <br>
        <br/>
        پاسخ:
        بله این دو مستقل‌اند . زیرا اگر بدانیم که رویداد C رخ داده تغییری در احتمالات تاس اول نمی‌گذارد و هر عددی می‌تواند آمده باشد.
        <br>
        <br/>
	</font>
</div>

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=blue size=6>
تعریف علمی مستقل بودن:
        </font><br/>
        دو رویداد A و B را در صورتی مستقل می‌نامیم که :$$P(A \cap B) = P(A)P(B)$$
        و این یعنی :
        $$P(A) = \frac{(A \cap B)}{P(B)}=P(A|B) $$
        <br/>
به زبان دیگر دو رویداد مستقل‌اند اگر و تنها اگر: $$P(A | B) = P(A)$$
    </font>
</div>

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=darkpink size=6>
تمرین:
        </font>
        نشان دهید دو رویداد $A$ و $B$ مستقل‌اند اگر و فقط اگر 
        $$.P(A|B) = P(A|\overline{B})$$
        تعبیر این عبارت ریاضی به زبان طبیعی چیست؟
</font>
</div>

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=darkpink size=6>
تمرین:
        </font>
گزاره های زیر را ثابت کنید:
        <br>
        <br/>
        ۱) اگر وقایع $A$ و $B$ مستقل باشند، نقیض $A$ و $B$ نیز مستقلند.
        <br>
        <br/>
        ۲)اگر $A$ و $B$ مستقل باشند، نقیض $A$ و نقیض $B$ نیز مستقلند.
</font>
</div>

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#682C94 size=5>
کنجکاوی:
        </font>
        اگر واقعه $A$ از واقعه $C$ و واقعه $B$ مستقل باشد، آیا می توان گفت که واقعه $A$ از واقعه $B \cap C$ مستقل است؟ به مفهوم این گزاره خوب دقت کنید. آیا می توان تعریفی از استقلال سه واقعه ارائه داد؟
	</font>
</div>


<p></p>
<br />
<div id="sec_problem" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
بازگشت به مسأله تاس
        </font>
		<p></p>
		<hr>
        می‌خواهیم جواب <a href="#taas"> مساله تاس </a> را به صورت علمی بیان کنیم:
        $$P(A\cap B) = P({4,2}) =\frac{1}{6} * \frac{1}{6} = \frac{1}{36}$$
        $$P(A)P(B) =\frac{1}{6} * \frac{5}{36} = \frac{5}{216}$$
        چون این دو مقدار برابر نیستند پس وابسته‌اند.
		<hr>
        $$P(A\cap C) = P({4,3}) =\frac{1}{6} * \frac{1}{6} = \frac{1}{36}$$
        $$P(A)P(C) =\frac{1}{6} * \frac{6}{36} = \frac{1}{36}$$
        چون این دو مقدار برابر‌اند پس مستقل‌اند.
	</font>
</div>

استقلال شرطی
        </font>
		<p></p>
		<hr>
        <font color=#2C5494 size=5>
مثال ۱ 
        </font>
        <br>
            یک دستگاه را در نظر بگیرید که در صورت وجود دود در یک اتاق هشدار می‌دهد.حال رویداد A را هشدار دادن دستگاه , رویداد B را وجود آتش در اتاق و رویداد C را وجود دود در اتاق در نظر بگیرید.
        <br>
        </br>
        <br>
        </br>
         آیا رویداد A و B 
        مستقل‌اند؟
        <br>
        </br>
        پاسخ: خیر . زیرا در صورت دانستن رخداد هر کدام , احتمال رخ دادن دیگری افزایش پیدا می‌کند.
        <br>
        </br>
        <br>
        </br>
          فرض کنید به ما گفته شود که رویداد C رخ داده است. حال آیا رویداد A و B مستقل‌اند؟
        <br>
        </br>
        پاسخ: بله . زیرا در صورت دانستن رخداد هر کدام , احتمال رخ دادن دیگری تغییری نمی‌کند.
        <br>
        </br>
        <br>
        <font color=#2C5494 size=5>
مثال ۲ 
        </font>
        <br>
            یک بازی را در نظر بگیرید که دو تاس دارد.اگر جمع دو تاس بیشتر از ۱۰ باشد , می‌بریم و در غیر این صورت می‌بازیم. حال رویداد A را 5 آمدن تاس اول و B را 5 آمدن تاس دوم در نظر بگیرید. همچنین رویداد C را برنده شدن بازیکن فرض می‌کنیم.
        <br>
        </br>
        <br>
        </br>
         آیا رویداد A و B 
        مستقل‌اند؟
        <br>
        </br>
        پاسخ: بله . زیرا در صورت دانستن رخداد هر کدام , احتمال رخ دادن دیگری تغییری نمی‌کند.
        <br>
        </br>
        <br>
        </br>
          فرض کنید به ما گفته شود که رویداد C رخ داده است. حال آیا رویداد A و B مستقل‌اند؟
        <br>
        </br>
        پاسخ: خیر . زیرا در صورت دانستن رخداد هر کدام , احتمال رخ دادن دیگری تغییر می‌کند.
        <br>
        </br>
        <br>
        </br>
        <font color=blue size=6>
تعریف علمی مستقل شرطی بودن:
        </font><br/>
    دو رویداد A و B را به صورت شرطی مستقل می‌نامیم اگر و تنها اگر رویداد C را به نحوی داشته باشیم که:
    $$P(A|BC) = P(A|C) $$

        </br>
        یا به عبارت دیگر داشته باشیم:
        $$P(AB|C) = P(A|C) * P(B|C) $$
        <br>
        </br> 
        در واقع حالت‌هایی که رویداد‌های a , c مستقل شرطی‌اند ( در صورت دانستن رویداد b ) را می‌توان به صورت زیر بیان کرد:

قانون زنجیره‌ای
        </font>
		<p></p>
		<hr>
         به طور کلی می‌دانیم که احتمال شرطی برای دو متغیر به شکل زیر تعریف می‌شود:
        $$P(A|B) = \frac{P(A \cap B)}{P(B)} $$
        <br>
        </br>
        حال قانون زنجیره‌ای بیان می‌دارد : $$P(A_1 \cap A_2 \cap ... \cap A_n) = P(A_1) * P(A_2|A_1) * P(A_3|A_2 \cap A_1) * ... * P(A_n|A_{n-1} \cap ... \cap A_2 \cap A_1)$$ 
        <br>
        </br> 
        <br>
        <br>
        <font color=#2C5494 size=5>
مثال 
        </font>
        <br>
        </br>
        فرض کنید دو ظرف داریم که در هر کدام یک تاس قرمز و یک تاس سبز موجود است. حال احتمال این که تاس قرمز از ظرف اول انتخاب شود و پس از پرتاب عدد ۴ را به ما نشان دهد چقدر است؟
        <br>
        </br>
        <br>
        </br>
        پاسخ: در این سوال اگر انتخاب ظرف اول را رویداد A و انتخاب تاس قرمز را رویداد B و آمدن عدد ۴ پس از پرتاب تاس را رویداد C تعریف کنیم , مطابق قانون زنجیره‌ای داریم:
        $$P(A \cap B \cap C) = P(A) * P(B|A) * P(C|B \cap A)$$
        $$P(A \cap B) = \frac{1}{2} * \frac{1}{2} = \frac{1}{6} = \frac{1}{24} $$
        <br>
        </br>
        <font color=#682C94 size=6>
کنج‌کاوی:
        </font>
        <br>
        </br>
        قانون زنجیره‌ای را با استفاده از تعریف احتمال شرطی که در بالا بیان شد اثبات کنید.
        <br>
        </br>
    </font>

        <ul>
            <li>

            </li>
