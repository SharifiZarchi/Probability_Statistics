        $$
        \begin{aligned}
        \end{aligned}
        $$
        بنابراین مسافت طی شده توسط ماشین‌ها اهمیتی ندارد و بر احتمال تمام شدن باتری در طی مسافت $s$ تاثیرگذار نیست. در نتیجه بهترین انتخاب، باتری با میانگین طول عمر بیشتر است هر چند که بیشتر از همه کار کرده باشد.

        پ.ن:در دنیای واقعی عمر باتری ماشین از توزیع نمایی پیروی نکرده و با گذر زمان احتمال بقای آن کمتر میشود اما از مثال‌های بی‌حافظگی میتوان به نیمه عمر مواد رادیواکتیو ،زمان کشف بلاک جدید بیت‌کوین و یا زمان رسیدن مشتری بعدی در نظریه‌ی صف اشاره کرد.
