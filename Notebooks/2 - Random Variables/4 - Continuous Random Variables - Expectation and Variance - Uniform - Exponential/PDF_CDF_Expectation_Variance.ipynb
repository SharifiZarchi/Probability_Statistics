            مقدمه
        </font>
		<p></p>
		<hr>
        در قسمت‌های قبلی با مفهوم متغیر تصادفی و متغیر تصادفی‌های گسسته و امید ریاضی، واریانس و خواص مربوط به آن‌ها آشنا شدیم. در این بخش، می‌خواهیم با دسته‌ی جدیدی از متغیرهای تصادفی کار کنیم.
        <br>
        سرعت یک ماشین در بزرگراه را در نظر گیرید. می‌دانیم سرعت آن با یک سرعت‌سنج سنجیده می‌شود که اعداد نمایش‌داده شده روی آن مقادیری گسسته دارند. اما اگر بخواهیم سرعت را دقیق‌تر بررسی کنیم، می‌بینیم که در واقع مقادیر پیوسته‌ای خواهد داشت.  بنابراین می‌توان این حرکت با با یک مقدار پیوسته نیز مدل کرد.
     <br>
        حال این مدل کردن به‌طور پیوسته، در بسیاری از مسائل کابرد خواهد داشت. علاوه بر این‌که در این مسائل مدل کردن با یک متغیر پیوسته، احتمالاً به واقعیت نزدیک‌تر خواهد بود، مزایای دیگری مانند استفاده از ابزارهایی که در درس‌های دیگر مانند ریاضی ۱ و ۲ آموخته‌ایم نیز وجود خواهد داشت تا با استفاده از این ابزارها، بتوانیم این مسائل را بهتر تحلیل کنیم.
<br>
        حال همان‌طور که در بخش‌های قبلی متغیرهای تصادفی پیوسته را بررسی کردیم، می‌خواهیم متغیرهای تصادفی گسسته را بررسی کنیم. خواهیم دید شباهت‌های بسیار زیادی بین این دو وجود دارد.

<div id="sec_pdf" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            متغیرهای تصادفی پیوسته و تابع چگالی احتمال
        </font>
        <hr>
        متغیر تصادفی $X$ را
        <strong>
            پیوسته
            </strong>
        می‌گوییم، اگر بتوان احتمال رخداد پیشامدهای مختلف آن را با استفاده از تابع نامنفی
        $f_X$
        به‌صورت زیر نشان داد:
        $$\mathbb{P} (X \in A) = \int_A f_X(x)dx$$
        که در آن $A$ هر زیرمجموعه از اعداد حقیقی است.
        و $f_X$ را
        تابع چگالی احتمال
       (Probability density function)
       یا به اختصار PDF می‌نامیم.
       <br>
        به‌طور خاص اگر بازه‌ی 
        $I = [a,b]$
        را در نظر گیریم خواهیم داشت:
        $$\mathbb{P} (X \in I) = \mathbb{P} (a \le X \le b) = \int_{a}^{b} f_X(x)dx$$
        <img src="figs/pdf_samplespace.png" style="width:80%"/>
        <br><br>
        <font color="blue">
            نکات مهم:
        </font>
        <ol>
        <li>
        حال می‌دانیم برای نقطه‌ی $a$ به‌طور خاص خواهیم داشت:
        $$\mathbb{P} (X = a) = \int_{a}^{a} f_X(x)dx  = 0$$
        بنابراین برای محاسبه‌ی احتمال بازه‌ها، باز یا بسته بودن آن‌ها برای ما اهمیتی نخواهد داشت و داریم:
        $$\mathbb{P} (a \le X \le b) = \mathbb{P} (a < X < b) = \mathbb{P} (a \le X < b) = \mathbb{P} (a < X \le b)$$
        </li>
        <br>
        <li>
        تابع جرمی احتمال و خواص آن را در نظر گیرید. می‌توان به همان ترتیب که در جزوه‌های قبلی آمده استِ، آن خواص را برای تابع چگالی نیز ثابت کرد. 
        <br>
        می‌دانیم تابع توزیع احتمال، تابعی است نامنفی. یعنی:
        برای هر $x$ داریم
        $f_X(x) \ge 0$
        .
       همچنین می‌دانیم خاصیت زیر را برای توابع چگالی احتمال برقرار است. 
       $$ \mathbb{P} (-\infty \le X \le \infty) = 1
       \Rightarrow
       \int_{-\infty}^{\infty} f_X(x)dx = 1 $$
        <br>
        <li>
        با توجه به قسمت ۱، نمی‌توانیم احتمال در یک نقطه را به‌دست بیاوریم. ولی اگر بخواهیم تقریبی از احتمال در آن نقطه داشته باشیم، می‌توان به این روش عمل کرد که بازه‌ای حول آن نقطه در نظر گرفت، و بعد احتمال آن بازه را به‌دست آورد. یعنی خواهیم داشت:
        $$\mathbb{P} (X \in [x, x+\delta]) = \mathbb{P} (x \le X \le x+\delta) = \int_{x}^{x+\delta} f_X(t)dt \approx
        f_X(x).\delta
        $$
        <img src="figs/pdf_delta.png" style="width:30%"/>
        <br>
        </ol>
        <br>
        <font color="blue">
        مثال:
        </font>
        تابع زیر را به‌عنوان تابع توزیع چگالی متغیر تصادفی $X$ تعریف کرده‌ایم
  .
  آیا این تابع معتبر است یا خیر؟ 
        $$
        \left\{
	\begin{array}{ll}
		\frac{1}{2\sqrt x}  & \mbox{if } 0 < x \le 1 \\
		0 & \mbox{ otherwise}
	\end{array}
        \right.
        $$
        <br>
        حل: در ابتدا ممکن است به نظر رسد که چون مقادیر $x$ در نزدیکی ۰ به بینهایت میل می‌کنند، این تابع معتبر نخواهد بود.
        اما برای بررسی آن، تنها کافی است شرط‌های بالا برقرار باشند.
        در این‌جا تنها شرط نابدیهی را چک می‌کنیم:
       $$ \mathbb{P} (-\infty \le X \le \infty) = 1
       \Rightarrow
       \int_{-\infty}^{\infty} f_X(x)dx = 
       \int_{0}^{1} \frac{1}{2\sqrt x}dx = 
       \sqrt x \big |^1_0 =
       1 $$
</font>
</div>

<div id="sec_cdf" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
تابع توزیع تجمعی
        </font>
		<hr>
همان‌طور که قبلاً دیده‌ایم، تابع توزیع انباشته یا تابع توزیع تجمعی
(Cumulative Distribution Function)
به صورت زیر تعریف می‌شود:
$$
F(x)=P(X\leq x)
$$
<br/>
همچنین می‌دانیم
$F(a)$
برابر مقدار
احتمال در بازه‌ی 
$[-\infty, a]$
است.
<br>
حال اگر $X$ یک متغیر تصادفی پیوسته باشد، خواهیم داشت:
$$ F(a)= \int_{-\infty}^{a} f_X(x) dx $$
<br>
	</font>
</div>

<div id="sec_attr" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
خواص تابع توزیع تجمعی
        </font>
		<p></p>
		<hr>
     در ادامه به تعدادی از خواص تابع توزیع انباشته اشاره می‌شود:
<ol>
    <li>
$$
\lim_{x\to-\infty}F(x)=0
$$
زیرا:
$$
F(-\infty)=P(\{X\leq-\infty\})=0
$$
    </li>
    <br />
    <li>
        $$
        \lim_{x\to\infty}F(x)=1
        $$
        زیرا:
        $$F(+\infty)=P(\{X\leq+\infty\})=1$$
    </li>
    <br />
    <li>
        $$x_1\leq x_2\Rightarrow F(x_1)\le F(x_2)$$
        این ویژگی به این معناست که $F(x)$ تابعی غیرنزولی است. زیرا:
        $$x_1\leq x_2\Rightarrow\{\omega:X(\omega)\leq x_1\}\subseteq\{\omega:X(\omega)\leq x_2\}\Rightarrow P(\{X\le x_1\})\leq P(\{X\leq x_2\})\Rightarrow F(x_1)\leq F(x_2)$$
به عبارت دیگر تابع توزیع تجمعی به این دلیل که برابر جمع مقادیر مثبت است، هیچ‌گاه کاهش نمی‌یابد.
<br/>
        همچنین از این جا نتیجه می‌شود که: $0\leq F(x)\leq1$، چون دیدیم که در $-\infty$ و $+\infty$ مقدار تابع $F$ به ترتیب صفر و یک است.
    </li>
    <br />
    <li>
    $$P(\{X>x\})=1-F(x)$$
    زیرا:
    $$\{\omega:X(\omega)>x\}=\{\omega:X(\omega)\le x\}^c\Rightarrow P(\{X>x\})=1-P(\{X\le x\})=1-F(x)$$
    </li>
    <br />
    <li>
    $$P(\{x_1<x\le x_2\})=F(x_2)-F(x_1)$$
زیرا:
<br/>
دو مجموعه
$A = \{X\le x_1\}$
و
$B = \{x_1<X\le x_2\}$
را در نظر بگیرید.
توجه کنید که
$A \cup B = \{x_1<x\le x_2\}$
و
$A \cap B = \emptyset$.
<br/>
    $$\{X\le x_2\}=A\cup B \Rightarrow P(\{X\le x_2\})=P(\{X\le x_1\})+P(\{x_1<X\le x_2\})\Rightarrow P(\{x_1<X\le x_2\})=F(x_2)-F(x_1)$$
    </li>
</ol>
	</font>
</div>

تعریف امید ریاضی پیوسته
        </font>
        		<hr>
		<br>
برای متغیر تصادفی گسسته
        $X$
امید ریاضی به صورت زیر تعریف می‌شود:
$$
\mathop{\mathbb{E}}[X] = \int_{-\infty}^{\infty}xf(x)dx
$$
که در آن
$f(x)$
تابع توزیع چگالی متغیر تصادفی
$X$
است.
<br>
<br>
همان‌طور که در بخش‌های قبلی دیدیم، برای متغیرهای تصادفی گسسته داریم:
$$\mathop{\mathbb{E}}[X] = \sum_{x:p(x)>0}{xp(x)}$$
که در آن
$p$
تابع توزیع چگالی متغیر تصادفی
$X$
است.
<br> 
<br>
حال با توجه به توضیحاتی که در قسمت تابع توزیع چگالی داده شد، توضیح دهید چرا امید ریاضی پیوسته را به‌صورت بالا تعریف کرده‌ایم. 
	</font>
</div>

فرض کنید که تابع چگالی احتمال متغیر تصادفی

$$\mathop{\mathbb{E}}[X] = \int_{-\infty}^{\infty}xf(x)dx$$

<div id="sec_sum_example" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<p></p>
        <font color="blue">
            مثال:
    </font>
        <font color=#FF7500 size=6>
            توزیع یکنواخت
        </font>
		<hr>
        آزمایشی را در نظر گیرید که در آن می‌خواهیم نقطه‌ای از بازه‌ی 
        $[a, b]$
        به‌دست بیاوریم به‌گونه‌ای که همه‌ی نقطه‌ها احتمال یکسانی داشته باشند.
        <br>
        یعنی به‌ازای هر دو نقطه مانند 
        $x_1, x_2 \in [a, b]$
        خواهیم داشت:
        $\mathbb{P} (X = x_1) = \mathbb{P} (X = x_2)$ .
        <br>
       که در این‌جا همان‌طور که در بالاتر توضیح داده شد، منظور بازه‌ای حول نقطه‌ی مورد نظر خواهد بود.
       <br>
       بنابراین، در آن بازه‌ی خاص باید داشته باشیم:
       $$f_X(x) = c \;\;\;\;\; x \in [a,b]  $$
       با توجه به خاصیت دوم تابع توزیع چگالی، این تابع به صورت زیر درخواهد آمد:
       $$   
f(x) = 
     \begin{cases}
       \frac{1}{b-a} &\quad x\in [a,b]\\
       \text{0} &\quad\text{otherwise} \\ 
     \end{cases}
$$
<br>
حال می‌خواهیم امید ریاضی این متغیر تصادفی را محاسبه کنیم.
<br>
$$ \begin{align*}
 \mathbb{E}[X] &= 
 \int_{-\infty}^{\infty} xf_X(x) \: dx \\
 &=
 \int_{a}^{b} x \cdot \frac{1}{b-a} \: dx \\
 &= 
 \frac{1}{b-a} \cdot \frac{1}{2} x^2 \Big|^b_a \\
 &= 
 \frac{1}{b-a} \cdot \frac{b^2 - a^2}{2} \\
 &= \frac{a+b}{2},
\end{align*}
$$
</font>
</div>

<div id="sec_function" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
امید ریاضی تابعی از متغیر تصادفی
        </font>
		<p></p>
		<hr>
        تابعی از یک متغیر تصادفی خود یک متغیر تصادفی است.
 در برخی از کاربردها نیاز به محاسبه امید ریاضی تابعی از یک متغیر تصادفی است. به طور مثال محاسبه‌ی امید ریاضی مربع یک متغیر تصادفی
از مواردی است که در درس به زودی با آن مواجه خواهید شد.
<br/>
نکته‌ای که باید به آن توجه کرد این است که امید ریاضی تابعی از یک متغیر لزوماً برابر با اعمال آن تابع بر روی امید ریاضی نیست.
<br/>
این در حالی است که امید ریاضی متغیر تصادفی
در حالت کلی امید ریاضی تابع
$g$
از متغیر تصادفی
X
برابر است با
$$
\mathop{\mathbb{E}}[g(X)] = \int_{-\infty}^{\infty}{g(x)\; f(x)dx}
$$
<br/>
      ممکن است عبارت بالا بدیهی بنظر برسد و از آن بارها استفاده کرده باشید (برای مثال زمانی که می‌خواهید
        $E[X^2]$
        را محاسبه کنید تا واریانس را بدست بیاورید.)
        اما این رابطه به هیج عنوان بدیهی نیست و نیاز به اثبات دارد!
        <br>
        از همین رو نیز به آن
        Law of the unconscious statistician
        یا به اختصار LOTUS گفته می‌شود.
        قضیه LOTUS برای هر نوع تابع g برقرار است و در اینجا حالت خاصی از آن که g
        تابعی یکنوا و مشتق‌پذیر است را ثابت می‌کنیم.
        <br>
        فرض کنید
        $Y = g(X)$ و
        $g^{-1}$
        معکوس تابع
        $g$ باشد. داریم:
        $$\frac{\mathrm{d}}{\mathrm{d}y} (g^{-1}(y)) = \frac{1}{g'(g^{-1}(y))}$$
        از آنجایی که
        $x = g^{-1}(y)$
        نتیجه می‌شود:
        $$\mathrm{d}x = \frac{1}{g'(g^{-1}(y))} \, \mathrm{d}y$$
        با تغییر متغیر خواهیم داشت:
        $$\int_{-\infty}^{\infty} g(x) f_X(x) \, \mathrm{d}x = \int_{-\infty}^{\infty} y \, f_X(g^{-1}(y)) \, \frac{1}{g'(g^{-1}(y))} \, \mathrm{d}y \;$$
        <br>
        حال تنها باید ثابت کنیم که
        $f_X(g^{-1}(y)) \, \frac{1}{g'(g^{-1}(y))} = f_Y(y)$
        <ul>
            <li> 
            اگر g صعودی باشد داریم:
$$\begin{align*}
F_Y(y) &= Pr(Y \leq y) \\
&= Pr(g(X) \leq y) \\
&= Pr(X \leq g^{-1}(y)) \\
&= F_X(g^{-1}(y)) \;
\end{align*}$$
                از طرفین مشتق می‌گیریم:
$$\begin{align*}
f_Y(y) &= \frac{\mathrm{d}}{\mathrm{d}y} F_X(g^{-1}(y)) \\
&= f_X(g^{-1}(y)) \, \frac{\mathrm{d}}{\mathrm{d}y} (g^{-1}(y)) \\
&= f_X(g^{-1}(y)) \, \frac{1}{g'(g^{-1}(y))} \;
\end{align*}$$
                </li>  
            <br>
            <li>
                اگر g نزولی باشد و 
                $h(X) = -g(X)$
                باشد:
$$\begin{align*}
F_Y(y) &= Pr(Y \leq y) \\
&= Pr(g(X) \leq y) \\
&= Pr(-g(X) \geq -y) \\
&= Pr(h(X) \geq -y) \\
&= Pr(X \geq h^{-1}(-y)) \\
&= 1 - F_X(h^{-1}(-y)) \;
\end{align*}$$
                با مشتق‌گیری از طرفین داریم:
$$\begin{align*}
f_Y(y) &= -f_X(h^{-1}(-y)) \, \frac{1}{h'(h^{-1}(-y))} \\
&= -f_X(h^{-1}(-y)) \, \frac{1}{-g'(h^{-1}(-y))} \\
&= f_X(h^{-1}(-y)) \, \frac{1}{g'(h^{-1}(-y))} \;
\end{align*}$$
                <br>
                از طرفی
                $h^{-1}(-y)$
                برابر $x$ ای است که:
                $h(x) = -g(x) = -y$
                در نتیجه:
                $$g(x) = y \Rightarrow x = g^{-1}(y) = h^{-1}(-y)$$
                و
                $$f_Y(y) = f_X(g^{-1}(y)) \, \frac{1}{g'(g^{-1}(y))}$$
            </li>
        </ul>
        <br>
                و با جایگذاری $f_Y(y)$ نتیجه می‌شود:
                $$\int_{-\infty}^{\infty} g(x) f_X(x) \, \mathrm{d}x = \int_{-\infty}^{\infty} y \, f_Y(y) \, \mathrm{d}y = E[Y] = E[g(X)] \;$$
	</font>
</div>

		<font color="blue">
            مثال:
    </font>
تکه چوبی به طول ۱ واحد داریم. این تکه چوب را از نقطه
U
می‌شکنیم. تابع توزیع احتمال
U
به صورت یکنواخت
است یعنی:
$$
f(u) = \begin{cases}
       1 &\quad u\in [0,1]\\
       0 &\quad\text{otherwise} \\ 
     \end{cases}
$$
نقطه
$0 \leq p \leq 1, p$
به ما داده شده است.
امید ریاضی طول قطعه چوبی را که شامل نقطه
$p$
است به دست آورید.
<br/>
<img src="figs/Stick_Function_Example.png" style="float:center; width:35%"/>
<br />
تابع
$L_p(U)$
از متغیر تصادفی
U
را تعریف می‌کنیم:
$$
L_p(U) = \begin{cases}
       1 - U &\quad U < p\\
       U &\quad U > p \\ 
     \end{cases}
$$
حال برای به دست آوردن امید ریاضی این تابع داریم:
$$
\begin{aligned}
\mathop{\mathbb{E}}[L_p(U)] &= \int_{-\infty}^{\infty}{L_p(u)f(u)du} \\
&= \int_{0}^{1}{L_p(u).1.du} \\
&= \int_{0}^{p}{(1-u)du} + \int_{p}^{1}{udu}\\
&= p(1-p) + \frac{1}{2}
\end{aligned}
$$
	</font>
</div>

<div id="sec_definition_example_discrete" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color="blue">
مثال:
        </font>
تکه چوبی به طول ۱ واحد داریم.
فرض کنید یک سر این چوب منطبق بر عدد
0
محور افقی و سر دیگر آن منطبق بر عدد
1
محور می‌باشد.
        این تکه چوب را از نقطه
$\frac{1}{2^U}$
می‌شکنیم که
U
متغیر تصادفی‌ای با تابع جرم احتمال به صورت زیر است
:
$$
P_U(u) = \begin{cases}
       \frac{1}{2^u} &\quad u\in \mathop{\mathbb{N}}\\
       0 &\quad\text{otherwise} \\ 
     \end{cases}
$$
عدد ثابت
$p$
که
$p \in \mathop{\mathbb{N}}$
است،
به ما داده شده است.
امید ریاضی طول قطعه چوبی را که شامل نقطه
$\frac{1}{2^p}$
بر روی محور است،
به دست آورید.
به شکل زیر توجه کنید.
<br/>
<img src="figs/wood.png" style="float:center; width:35%"/>
در حالت تساوی دو عدد
$p$
و
$u$
قطعه چوبی را که یک انتهای آن عدد صفر محور است،
در نظر بگیرید.
<br/>
<br/>
تابع
$L_p(U)$
از متغیر تصادفی
U
را تعریف می‌کنیم. این تابع خود یک متغیر تصادفی خواهد بود و امید ریاضی آن پاسخ به سوال است
$$
L_p(U) = \begin{cases}
       \frac{1}{2^U} &\quad U \leq p\\
       1 - \frac{1}{2^U} &\quad U > p \\ 
     \end{cases}
$$
حال برای به دست آوردن امید ریاضی این تابع داریم:
$$
\begin{aligned}
\mathop{\mathbb{E}}[L_p(U)] &= \sum_{u=1}^{\infty}{L_p(u)P_U(U=u)} \\
&= \sum_{u=1}^{p}{(\frac{1}{2^u}).P_U(U=u)} + \sum_{u=p+1}^{\infty}{(1 - \frac{1}{2^u}).P_U(U=u)}\\
&= \sum_{u=1}^{p}{(\frac{1}{2^u})(\frac{1}{2^u})} + \sum_{u=p+1}^{\infty}{(1 - \frac{1}{2^u})(\frac{1}{2^u})}\\
&= \frac{1}{3}(1 - \frac{1}{4^{p+1}}) + \frac{1}{2^p} - \frac{1}{3}(\frac{1}{4^p})
\end{aligned}
$$

خطی بودن امید ریاضی و جمع متغیرهای تصادفی
        </font>
		<p></p>
		<hr>
.یکی از ویژگی‌های مهم امید ریاضی خطی بودن آن است
$$
\mathop{\mathbb{E}}[aX + b] = a\mathop{\mathbb{E}}[X] + b
$$
اثبات رابطه بالا به کمک قسمت قبل به دست می‌آید.
برای این کار تابع زیر را تعریف می‌کنیم:
$$
g(X) = aX + b
$$
سپس با توجه به قسمت قبل داریم:
$$
\begin{aligned}
\mathop{\mathbb{E}}[aX + b] &= \mathop{\mathbb{E}}[g(X)] \\
&= \int_{-\infty}^{+\infty} (ax+b)f_X(x)dx \\
&= \int_{-\infty}^{+\infty} axf_X(x)dx + 
\int_{-\infty}^{+\infty} bf_X(x)dx\\
&= a \int_{-\infty}^{+\infty} xf_X(x)dx + 
b \int_{-\infty}^{+\infty} f_X(x)dx \\
&= a\mathop{\mathbb{E}}[X] + b
\end{aligned}
$$
 به طور خاص استفاده از این ویژگی و توابع نشانگر
می‌تواند حل برخی از مسائل را بسیار ساده‌تر کند.
<br/>
<br/>
دیدیم که ضرب عدد ثابت در متغیر تصادفی یا جمع آن با عدد ثابت،
مقدار امید ریاضی را به همان میزان جابجا کرد.
حال سوال بعدی این است که اگر به جای عدد ثابت،
متغیر تصادفی را یک متغیر تصادفی دیگر جمع کنیم،
چه اتفاقی برای امید ریاضی آن‌ها می‌افتد؟
به عبارت دقیق تر اگر
$Z = X + Y$
باشد،
آیا
گزاره
$$
\mathop{\mathbb{E}}[Z] = \mathop{\mathbb{E}}[X + Y] = \mathop{\mathbb{E}}[X] + \mathop{\mathbb{E}}[Y]
$$
برقرار است؟
<br/>
<br/>
پاسخ مثبت است. اثبات به کمک استفاده از توزیع مشترک دو متغیر
$X$
و
$Y$
:
$$
\begin{aligned}
\mathop{\mathbb{E}}[Z] &= \mathop{\mathbb{E}}[(X + Y)]\\
&= \int_x\int_y(x+y)f_{XY}(x,y)dydx \\
&= \int_x\int_yxf_{XY}(x,y)dydx + \int_y\int_xyf_{XY}(x,y)dxdy \\
&= \int_xxf_{X}(x)dx + \int_yyf_{Y}(y)dy\\
&= \mathop{\mathbb{E}}[X] + \mathop{\mathbb{E}}[Y]
\end{aligned}
$$
همان طور که مشاهده می‌کنید در اثبات بالا در هیچ جا از فرض مستقل بودن یا نبودن دو متغیر تصادفی نسبت به یکدیگر استفاده نشده است.
یعنی این گزاره برای
<strong>
هر دو متغیر تصادفی‌ای، چه مستقل چه وابسته،
</strong>
برقرار است.
<br/>
<br/>
اثبات گفته شده در بالا کمی نقص دارد. در آن به نکته‌ای اشاره نشده است. آیا می‌توانید آن را پیدا کنید؟ 
	</font>
</div>

<div id="sec_def_variance" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
تعریف واریانس
        </font>
		<p></p>
		<hr>
       واریانس یک متغیر تصادفی 
را که معیاری از پراکندگی داده خروجی‌های ممکن آن است،
به شرح زیر تعریف کردیم:
$$
Var(X) = \mathop{\mathbb{E}}[(X - \mathop{\mathbb{E}[X]})^2]
$$
یا در برخی نمادگذاری‌های دیگر به صورت زیر تعریف می‌شود:
$$
Var(X) = \mathop{\mathbb{E}}[(X - \mu)^2]
$$
که در آن
$\mathop{\mathbb{E}}[X] = \mu$
است.
<br/>
با ساده کردن عبارت بالا داریم:
$$
\begin{aligned}
Var(X) &= \mathop{\mathbb{E}}[(X - \mu)^2] \\
&= \int_{-\infty}^{+\infty}(x-\mu)^2f_X(x) \\
&= \int_{-\infty}^{+\infty}(x^2+\mu^2-2\mu{x})f_X(x) \\
&= \int_{-\infty}^{+\infty} x^2f_X(x) + 
\int_{-\infty}^{+\infty}\mu^2f_X(x) -2\mu\int_{-\infty}^{+\infty} xf_X(x)\\
&= \mathop{\mathbb{E}}[X^2] + \mu^2.1 -2\mu.\mu \\
&= \mathop{\mathbb{E}}[X^2] - \mu^2 
\end{aligned} $$
       <br>
        پس برای واریانس یک متغیر تصادفی پیوسته خواهیم داشت:
$$ Var(X) = \mathop{\mathbb{E}}[X^2] -  (\mathop{\mathbb{E}}[X])^2
$$
<br/>
همچنین به جذر واریانس، انحراف معیار می‌گویند.
$$
SD(X) = \sqrt{Var(X)}
$$

</div>

        برای به‌دست آوردن واریانس توزیع یکنواخت به‌صورت زیر عمل می‌کنیم:
        $$
        \begin{align*}
 \mathbb{E}[X^2] &= 
 \int_{a}^{b}\frac{x^2}{b-a} \: dx \\
 &= 
 \frac{1}{b-a }\int_{a}^{b} x^2 \: dx \\
 &= 
 \frac{1}{b-a} \cdot \frac{1}{3} x^3 \Big|^b_a \\
 &= 
 \frac{b^3 - a^3}{3(b-a)} \\
 &= \frac{a^2+ab+b^2}{3}.
\end{align*}
$$
<br>
$$
\begin{equation*}
\text{var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 =
\frac{a^2+ab+b^2}{3} - \frac{(a+b)^2}{4} = \frac{(b-a)^2}{12},
\end{equation*}$$

<div id="sec_prop_variance" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
خواص واریانس
        </font>
		<p></p>
		<hr>
واریانس معیاری از پراکندگی خروجی‌های مختلف یک متغیر تصادفی است.
بنابراین انتقال یک متغیر تصادفی به اندازه یک عدد ثابت نباید اثری در آن داشته باشد چون پراکندگی نسبی خروجی‌های متغیر تصادفی انتقال‌یافته، نسبت به یکدیگر عوض نمی‌شود.
<br/>
همچنین واریانس با توان دوم مرتبط است. بنابراین دور از  
انتظار نیست که ضرب یک متغیر تصادفی در مقدار
ثابت رابطه توان دوم با واریانس متغیر اولیه داشته باشد.
<br/>
دو خاصیت بالا را در رابطه زیر می‌یابیم:
$$
\begin{aligned}
Var(aX + b) &= \mathop{\mathbb{E}}[(aX+b-a\mu-b)^2] \\
&= \mathop{\mathbb{E}}[a^2(X-\mu)^2] \\
&= a^2\mathop{\mathbb{E}}[(X-\mu)^2] \\
&= a^2Var(X)
\end{aligned}
$$
