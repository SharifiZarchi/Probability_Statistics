<div id="sec_intro_rv" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مقدمه
        </font>
		<p></p>
		<hr>
در قسمت‌های قبلی درس یاد گرفتیم که فضای نمونه
مانند
$\Omega$
مجموعه‌ای از همه خروجی‌های مختلف یک آزمایش تصادفی است.
        <br>
        متغیرهای تصادفی توابعی هستند که این خروجی‌ها را به اعداد حقیقی نگاشت می‌کنند.
<br/>
     این نگاشت در برخی آزمایش‌ها واضح‌تر از بقیه آزمایش‌هاست. به طور مثال در آزمایش تصادفی پرتاب تاس، خروجی‌های تاس اعداد ۱ تا ۶ هستند. بنابراین اولین گزینه‌ای که برای  تعریف  متغیر تصادفی
به ذهن می‌رسد نگاشت 
((عددی که پس از پرتاب روی تاس می‌آید))
به اعداد طبیعی بازه ۱ تا ۶ است.
        <br/>
        ولی در مورد پرتاب سکه این امر کمی متفاوت است. زیرا
        شیر یا خط آمدن سکه به طور ذاتی هیچ
        ((صفر))
        یا
        ((یک))
        ای
        در خود ندارد، ولی ما معمولاً بدون این‌که به این مورد زیاد فکر کنیم یکی از آن‌ها را به صفر و دیگری را به یک می‌نگاریم. این پل بین
        شیر یا خط و 0 یا ۱
        همان متغیر تصادفی است.
<br/>
<br/>
حال شاید این سوال ایجاد شود که چرا یک تابع را
((متغیر تصادفی))
می‌نامیم؟
        <br/>
        دلیل آن این است که
چون به خروجی یک آزمایش
تصادفی
        مربوط است،
مقدار آن
تصادفی
است و چون با آن‌ها مانند متغیرهای ریاضی
رفتار می‌کنیم و آن‌ها را با یکدیگر جمع و ضرب می‌کنیم، به توان می‌رسانیم و عملیات ریاضی انجام می‌دهیم، آن‌ها را متغیر نیز می‌نامیم.
<br />
	</font>

تعریف
            <p></p>
        </font>
        <hr>
        فرض کنید یک آزمایش و یک مجموعه از خروجی‌های ممکن (فضای نمونه) Omega در اختیار داشته باشیم. یک متغیر تصادفی، به هر یک از خروجی‌های ممکن مانندw  که عضو مجموعه Omega است یک عدد تخصیص می‌دهد. به بیان ریاضی می‌توان گفت               متغیر تصادفی تابعی است که یک فضای نمونه (مجموعه تمام نتایج یک آزمایش تصادفی) را به اعداد حقیقی نگاشت میکند. دامنهٔ این تابع مجموعهٔ Omega بوده و برد آن مجموعه‌ای از اعداد حقیقی است.
<img src="figs/rv.png" style="float:center; width:50%"/>
به طور مثال در آزمایش پرتاب تاس فضای نمونه ما 6 روی تاس است که هرکدام تعدادی دایره دارد, که از طریق متغیر تصادفی این تعداد دایره ها به اعداد حقیقی نگاشت میشوند.
<img src="figs/dice4_2.png" style="width:50%;" />
	</font>

<div id="sec_def_rv" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
متغیر های تصادفی گسسته و پیوسته
        </font>
		<p></p>
		<hr>
        متغیرهای تصادفی را به صورت کلی به دو نوع پیوسته و گسسته می‌توان تقسیم کرد.
<br/>
        متغیر تصادفی گسسته خروجی‌های مختلف یک آزمایش تصادفی را به زیر مجموعه‌ای شمارا (متناهی یا نامتناهی) از اعداد حقیقی نگاشت می‌کند. مثلاً خروجی یک بار پرتاب تاس اعداد طبیعی ۱ تا ۶ است یا
تعداد دفعاتی که بایستی یک سکه انداخته شود تا برای اولین بار شیر بیاید، مجموعه همه اعداد طبیعی است که شمارا است. بنابراین در این آزمایش‌ها متغیر تصادفی گسسته است.
<br/>
متغیرهای تصادفی پیوسته هم متغیرهایی هستند که خروجی یک آزمایش تصادفی را به زیرمجموعه‌ی نامتناهی ناشمارا از اعداد حقیقی نگاشت می‌کنند. این زیرمجموعه، که برد متغیر تصادفی است، به طور مثال می‌تواند
$[4,7]$،
$[0, \infty)$
یا
$\mathbb{R}$
باشد.
متغیر تصادفی پیوسته برای نگاشت خروجی‌های آزمایشی به کار می‌رود که
فضای نمونه آن
ناشمارا باشد.
معمولاً در این درس مجموعه ناشمارای مهمی که با آن سر و کار داریم، مجموعه اعداد حقیقی است.
یعنی به طور ساده می‌توان جمله بالا را این‌گونه تعبیر کرد که اگر آزمایش تصادفی ما خروجی‌ای حقیقی و پیوسته داشته باشد، متغیر تصادفی ما پیوسته است.
<br/>
        به طور مثال همه افراد زنده روی کره زمین در یک لحظه خاص
را در نظر بگیرید. فردی را به تصادف انتخاب می‌کنیم. اگر داده مورد علاقه ما سن افراد باشد، متغیر تصادفی‌ای که تعریف می‌کنیم به عنوان ورودی یک انسان می‌گیرد و به عنوان خروجی عدد سن فرد را فرضاً به سال تحویل می‌دهد
(مثلاً فردی می‌تواند
$12.3456$
سال سن داشته باشد).
        اگر فرض کنیم که انسان حداکثر ۱۵۰ سال می‌تواند عمر کند،
برد متغیر تصادفی، اعداد حقیقی بازه
$[0, 150]$
است.
این مجموعه ناشماراست، بنابراین متغیر تصادفی ما پیوسته است.
<br/>
(به بیان ساده اگر خروجی آزمایش از جنس گسسته مثل عدد تاس و شیر یا خط بودن سکه بود، متغیر تصادفی گسسته است. اگر از جنس پیوسته مثل طول و زمان بود، متغیر پیوسته است.)
<br/>
	</font>
</div>

چند مثال
        </font>
		<p></p>
		<hr>
        چند نمونه آزمایش‌های زیر را در نظر بگیرید:
<ul>
    <li>
        در یک آزمایش شامل پرتاب ۵ سکه، تعداد «شیر»ها در این دنباله یک متغیر تصادفی است. البته توجه کنید که خود دنباله یک متغیر تصادفی نیست چون یک مقدار عددی مشخص ندارد.
    </li>
    <li>
        در یک آزمایش ۲ بار یک تاس را پرتاب می‌کنیم. هر یک از موارد زیر یک متغیر تصادفی محسوب می‌شوند:
        <ol>
            <li>
                مجموع عدد حاصل از دو پرتاب
            </li>
            <li>
                تعداد ۶ها در دو پرتاب
            </li>
        </ol>
    </li>
    <li>
        در یک آزمایش شامل انتقال یک پیام، زمان لازم برای انتقال پیام، تعداد نشانه‌های اشتباه دریافت شده و تاخیر دریافت پیام همه متغیرهای تصادفی به حساب می‌آیند.
    </li>
</ul>
	</font>
</div>

	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
محاسبهٔ احتمال با استفاده از متغیر تصادفی
        </font>
		<p></p>
		<hr>
        آزمایش ۳ بار پرتاب یک سکهٔ سالم را در نظر بگیرید. فضای حالات ممکن بدین گونه است:
        <br/>
        $$\Omega=\{HHH, HHT, \dots, TTH, TTT\}$$
        حال فرض کنید متغیر تصادفی
        $X$
        را، تعداد شیرهای واقعهٔ
        $\omega$ تعریف کنیم.
        <br/>
        مقدار این متغیر تصادفی به این صورت خواهد بود:
        </font>
        </div>

    <th>THH</th> 
    <th>THT</th>

  </tr>

<div id="sec_probability2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
حال در این مثال احتمال این که $X(\omega)\le1$
باشد چیست؟
        <br/>
        احتمال روی همهٔ واقعه‌ها تعریف می‌شود. احتمال $X(\omega)\le1$ یعنی احتمال مجموعهٔ آن $\omega$هایی که برای آن‌ها $X(\omega)\le1$ باشد. یعنی:
$$P\{X\le1\}=P\{\omega:X(\omega)\le1\}=P\{HTT, THT, TTH, TTT\}=\frac{4}{8}$$
دقت کنید که معمولا به طور خلاصه می‌نویسیم: $\{X\le x\}$، ولی این مجموعه‌ای از اعداد نیست، بلکه یعنی: $\{\omega:X(\omega)\le x\}$.
<br/>
حال مثلا اگر احتمال $\{X\le x\}$ را برای هر $X$ بدانیم، احتمال همهٔ واقعه‌های مورد نظر را خواهیم داشت و دیگر لزومی به دانستن فضای احتمال $\Omega$ نیست.
	</font>
</div>

<div id="sec_intro_pmfcdf" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مقدمه
        </font>
		<p></p>
		<hr>
در بخش قبل با تعریف متغیر تصادفی و محاسبهٔ احتمال‌های مرتبط با آن آشنا شدیم و از عباراتی مانند
$P(X(\omega) = x)$
یا
$P(X(\omega) \leq x)$
استفاده کردیم.
        <br />
معنی عبارت
$P(X(\omega) = x)$
یعنی جمع احتمال وقوع همه رویدادهایی مانند
$\omega$
که
$\omega \in \Omega$
است و متغیر تصادفی آن
$\omega$
خاص را
به
$x$
می‌نگارد.
<br />
به عنوان مثال آزمایش تصادفی پرتاب دو تاس متقارن و سالم را در نظر بگیرید. متغیر تصادفی
$X$
را
این‌گونه تعریف می‌کنیم که
به ازای هر
$\omega=(d_1, d_2)$
که
$d_1$
و
$d_2$
عدد خروجی پرتاب تاس اول و دوم هستند،
$X(\omega) = max(d_1, d_2)$.
به عبارت دیگر این متغیر تصادفی برابر با عدد بیشینه هر بار پرتاب دو تاس است. یعنی اگر خروجی یک بار پرتاب همزمان دو تاس اعداد ۴ و ۵ باشند،
$X(\omega)=X((4,5)) = 5$.
<br />
حال فرض کنید که علاقه‌مند به دانستن
این هستیم که
اعدادی که در برد
$X$
قرار دارند
(اعداد طبیعی ۱ تا ۶)
هر کدام چه قدر احتمال وقوع دارند.
به عنوان
$P(X(\omega) = 2)$
را می‌خواهیم به دست آوریم.
برای این کار همان‌طور که در مثال جزوه قبلی دیدیم بایستی ابتدا همه
$\omega \in \Omega $
ها را در نظر بگیریم که
$X$
آن‌ها را به ۲ می‌نگارد.
یعنی:
$$\{(1,2), (2,1), (2,2)\}$$
سپس احتمال مد نظر، برابر جمع احتمال تک تک این رویدادهای مجزاست. در نتیجه:
$$
P(X(\omega) = 2) = \frac{1}{36} + \frac{1}{36} + \frac{1}{36} = \frac{3}{36}
$$
حال از اینجا به بعد به جای
$P(X(\omega)=x)$
یا
$P(X=x)$،
به طور خلاصه
از نماد
$p(x)$
استفاده می‌کنیم.
در مثال بالا
مقدار
$p(2)$
برابر
با
$\frac{3}{36}$
بود.
به این تابع،
تابع جرم احتمال می‌گوییم.
	</font>
</div>

<div id="sec_def_pmf" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
 تعریف تابع جرمی احتمال (PMF)
        </font>
		<p></p>
		<hr>
        فرض کنید
        $X$
یک متغیر تصادفی و
$x$
یکی از مقادیر ممکن آن باشد.تابع جرم احتمال
$X$
(Probability Mass Function)
که با
$p_X(a)$
نمایش داده می‌شود برابر با احتمال وقوع رویداد
$\{X=a\}$
است. یعنی:
        $$p_X(a)=P(\{X=a\})$$
مانند مثال بالا، اگر تنها با یک متغیر تصادفی سر و کار داشته باشیم، می‌توانیم اندیس $X$ را حذف کرده و $p_X(a)$ را به صورت $p(a)$ بنویسیم.
        <br />
توجه کنید که:
<ul>
<li>
    $\forall a \in \mathbb{R}: 0 \leq p(a) \leq 1$
</li>
<li>
$a$
هر عدد حقیقی‌ای می‌تواند باشد. در صورتی که
$a$
جز خروجی‌های ممکن
$X$
نباشد،
$p(a) = 0$
است.
</li>
</ul>
        <br />
در مثال قبلی، مقدار
$p(2)$
را محاسبه کردیم.
مقدار تابع جرم احتمال برای سایر مقادیر ممکن
$X$
در جدول زیر آمده است:
(تابع جرم احتمال به اختصار
pmf
نیز گفته می‌شود.
)
<table style="font-size: 0.9em;border:1px solid;">

تعریف تابع توزیع انباشته (CDF)
        </font>
		<p></p>
		<hr>
برابر قرار دادن یک متغیر تصادفی با یک عدد به مفهوم در نظر گرفتن رویدادهایی است که آن متغیر تصادفی به آن عدد می‌نگارد.
یعنی به طور مثال در بخش‌های قبلی
عبارت
$X=2$
برابر
$X(\omega)=2$
یا به بیان دیگر مجموعه رویدادهایی که متغیر تصادفی آن‌ها را به ۲
می‌نگارد، در نظر گرفته می‌شد.
حال به طریق مشابهی می‌توان از متغیرهای تصادفی در نامساوی ها 
(
یا سایر عبارات جبری
)
استفاده کرد. به طور نمونه 
$X(\omega) \leq a$
یعنی مجموعه رویدادهایی
مانند
$\omega$
که
مقدار
$X$
در آن‌ها
از
$a$
کوچک‌تر است.
<br/>
به طور مثال پرتاب همزمان دو تاس سالم را در نظر بگیرید. این‌بار متغیر تصادفی
$Z$
را
برابر جمع عدد روی دو تاس در نظر بگیرید.
$$
Z \leq 4 = \{(1,1), (1,2), (1,3), (2,1), (2,2), (3,1)\}
$$
با در نظر گرفتن این نامساوی، تابع توزیع انباشته را تعریف می‌کنیم.
<br/>
<br/>
تابع توزیع انباشته یا تابع توزیع تجمعی
(Cumulative Distribution Function)
به صورت زیر تعریف می‌شود:
$$
F(x)=P(X\leq x)
$$
<br/>
مثال پرتاب همزمان دو تاس و متغیر تصادفی بیشینه مقدار دو تاس را در نظر بگیرید.
مقادیر تابع توزیع انباشته
آن به شرح زیر است:

<div id="sec_prop_cdf" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
خواص تابع توزیع انباشته
        </font>
		<p></p>
		<hr>
     در ادامه به تعدادی از خواص تابع توزیع انباشته اشاره می‌شود:
<ol>
    <li>
$$
\lim_{x\to-\infty}F(x)=0
$$
زیرا:
$$
F(-\infty)=P(\{X\leq-\infty\})=0
$$
    </li>
    <br />
    <li>
        $$
        \lim_{x\to\infty}F(x)=1
        $$
        زیرا:
        $$F(+\infty)=P(\{X\leq+\infty\})=1$$
    </li>
    <br />
    <li>
        $$x_1\leq x_2\Rightarrow F(x_1)\le F(x_2)$$
        این ویژگی به این معناست که $F(x)$ تابعی غیرنزولی است. زیرا:
        $$x_1\leq x_2\Rightarrow\{\omega:X(\omega)\leq x_1\}\subseteq\{\omega:X(\omega)\leq x_2\}\Rightarrow P(\{X\le x_1\})\leq P(\{X\leq x_2\})\Rightarrow F(x_1)\leq F(x_2)$$
به عبارت دیگر تابع توزیع تجمعی به این دلیل که برابر جمع مقادیر مثبت است، هیچ‌گاه کاهش نمی‌یابد.
<br/>
        همچنین از این جا نتیجه می‌شود که: $0\leq F(x)\leq1$، چون دیدیم که در $-\infty$ و $+\infty$ مقدار تابع $F$ به ترتیب صفر و یک است.
    </li>
    <br />
    <li>
    $$P(\{X>x\})=1-F(x)$$
    زیرا:
    $$\{\omega:X(\omega)>x\}=\{\omega:X(\omega)\le x\}^c\Rightarrow P(\{X>x\})=1-P(\{X\le x\})=1-F(x)$$
    </li>
    <br />
    <li>
    $$P(\{x_1<x\le x_2\})=F(x_2)-F(x_1)$$
زیرا:
<br/>
دو مجموعه
$A = \{X\le x_1\}$
و
$B = \{x_1<X\le x_2\}$
را در نظر بگیرید.
توجه کنید که
$A \cup B = \{x_1<\le x_2\}$
و
$A \cap B = \emptyset$.
<br/>
    $$\{X\le x_2\}=A\cup B \Rightarrow P(\{X\le x_2\})=P(\{X\le x_1\})+P(\{x_1<X\le x_2\})\Rightarrow P(\{x_1<X\le x_2\})=F(x_2)-F(x_1)$$
    </li>
</ol>
	</font>
</div>

مقدمه
        </font>
		<p></p>
		<hr>
در قسمت‌های قبلی درس با احتمال و متغیرهای تصادفی آشنا شدیم  و در مسائل، احتمال وقوع رخدادهای مختلف را به صورت جداگانه حساب کردیم. حال سوالی که برای ذهن ایجاد می‌شود این است که
چگونه با دانستن احتمال وقوع رویدادهای مختلف و با ترکیب این اطلاعات با یک‌دیگر، می‌توان شهودی نسبت به نتیجه یک آزمایش تصادفی به دست آورد؟
        <br/>
        یکی از ساده‌ترین جواب‌ها به سوال بالا استفاده از مفهوم امید ریاضی است. به عبارت ساده و نادقیق، امید ریاضی میانگین وزن‌دار بر روی متغیر تصادفی است.
        <br/><br />
    </font>

<div id="sec_def_expected" style="direction:rtl;line-height:300%;">

تعریف امید ریاضی
        </font>
		<p></p>
		<hr>
برای متغیر تصادفی گسسته
        $X$
امید ریاضی به صورت زیر تعریف می‌شود:
$$\mathop{\mathbb{E}}[X] = \sum_{x:p(x)>0}{xp(x)}$$
که در آن
p
تابع جرم احتمال متغیر تصادفی
$X$
است.
به عبارت دیگر تعریف بالا معادل این است که بر روی تمام حالت‌های ممکن متغیر
$X$
میانگین وزن‌دار بر حسب میزان محتمل بودنشان گرفته شود.
<br/>
	</font>
</div>

<div id="sec_ex_expected" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مثال حالت گسسته
        </font>
		<p></p>
		<hr>
فرض کنید که یک تاس سالم با ۶ وجه داریم. اگر
  متغیر تصادفی
$X$
را برابر عدد خروجی پرتاب یک بار تاس تعریف کنیم،
امید ریاضی آن چند است؟
$$\mathop{\mathbb{E}}[X] = \sum_{x:p(x)>0}{xp(x)}$$
$$
\mathop{\mathbb{E}}[X] = 1(\frac{1}{6}) + 2(\frac{1}{6}) + 3(\frac{1}{6}) + 4(\frac{1}{6}) + 5(\frac{1}{6}) + 6(\frac{1}{6}) = \frac{7}{2}
$$

<br/>
برای مثالی دیگر میتوانیم به سکه اشاره کنیم.
به طور مثال، متغیر تصادفی
$X$
مقدار
۱
را در صورتی که نتیجه پرتاب سکه شیر شود، خروجی می‌دهد و در غیر این‌صورت مقدار آن
صفر
خواهد بود.     
حال اگر سکه متقارن باشد برای محاسبه میانگین وزن‌دار یا همان امید ریاضی در این حالت خواهیم داشت:
$$
\frac{1}{2}\times 1 + \frac{1}{2}\times 0 = \frac{1}{2}
$$
        
</font>
