<div id="sec_def_var" style="direction:rtl;line-height:300%;">

تعریف واریانس
        </font>
		<p></p>
		<hr>
       واریانس یک متغیر تصادفی گسسته
را که معیاری از پراکندگی داده خروجی‌های ممکن آن است،
به شرح زیر تعریف می‌کنیم:
$$
Var(X) = \mathop{\mathbb{E}}[(X - \mathop{\mathbb{E}[X]})^2]
$$
یا در برخی نمادگذاری‌های دیگر به صورت زیر تعریف می‌شود:
$$
Var(X) = \mathop{\mathbb{E}}[(X - \mu)^2]
$$
که در آن
$\mathop{\mathbb{E}}[X] = \mu$
است.
<br/>
با ساده کردن عبارت بالا داریم:
$$
\begin{aligned}
Var(X) &= \mathop{\mathbb{E}}[(X - \mu)^2] \\
&= \sum_{x}(x-\mu)^2p(x) \\
&= \sum_{x}(x^2+\mu^2-2\mu{x})p(x) \\
&= \sum_{x}x^2p(x) + \sum_{x}\mu^2p(x) -2\mu\sum_{x}xp(x)\\
&= \mathop{\mathbb{E}}[X^2] + \mu^2.1 -2\mu.\mu \\
&= \mathop{\mathbb{E}}[X^2] - \mu^2 \\
Var(X) &= \mathop{\mathbb{E}}[X^2] -  (\mathop{\mathbb{E}}[X])^2\\
\end{aligned}
$$
<br/>
همچنین به جذر واریانس، انحراف معیار می‌گویند.
$$
SD(X) = \sqrt{Var(X)}
$$
در برخی جاها واریانس دید بهتری میتواند به ما بدهد تا انحراف معیار و در برخی جاهای دیگر برعکس, به همین دلیل با اینکه با هم رابطه ی بسیار ساده ای دارند و با داشتن یکی دیگری را به سادگی میتوان به دست آورد, هردوی آن ها استفاده میشوند.
    </font>
</div>

<div id="sec_prop_var" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
خواص واریانس
        </font>
		<p></p>
		<hr>
واریانس معیاری از پراکندگی خروجی‌های مختلف یک متغیر تصادفی است.
بنابراین انتقال یک متغیر تصادفی به اندازه یک عدد ثابت نباید اثری در آن داشته باشد چون پراکندگی نسبی خروجی‌های متغیر تصادفی انتقال‌یافته، نسبت به یکدیگر عوض نمی‌شود.
<br/>
همچنین واریانس با توان دوم مرتبط است. بنابراین دور از  
انتظار نیست که ضرب یک متغیر تصادفی در مقدار
ثابت رابطه توان دوم با واریانس متغیر اولیه داشته باشد.
<br/>
دو خاصیت بالا را در رابطه زیر می‌یابیم:
$$
\begin{aligned}
Var(aX + b) &= \mathop{\mathbb{E}}[(aX+b-a\mu-b)^2] \\
&= \mathop{\mathbb{E}}[a^2(X-\mu)^2] \\
&= a^2\mathop{\mathbb{E}}[(X-\mu)^2] \\
&= a^2Var(X)
\end{aligned}
$$
اگر $X_1$،$X_2$،...،$X_n$ متغیرهای تصادفی دو به دو مستقل از هم باشند و $X=X_1+X_2+...+X_n$ باشد، داریم:
$$
\operatorname{Var}(X) = \operatorname{Var}(X_1)+\operatorname{Var}(X_2)+...+\operatorname{Var}(X_n)
$$
در آینده رابطه بالا را اثبات می‌کنیم.
    </font>
</div>

۱.۱

به دست می‌آوریم و اگر ناهم‌رنگ باشند ۱ دلار می‌بازیم.

$X$

\begin{aligned}

توزیع گسسته برنولی
        </font>
		<p></p>
		<hr>
        توزیع برنولی، توزیعی گسسته است که نام آن از نام دانشمند سوئیسی ژاکوب برنولی گرفته شده‌است. توزیع برنولی یک توزیع گسسته است که مقادیر یک (در صورت موفقیت آزمایش ) و صفر را (در صورت شکست) می‌گیرد.  احتمال موفقیت آزمایش برابر p است و احتمال شکست آن برابر q=1-p است. به عبارت دیگر توزیع برنولی به نوعی مربوط به یک سوال بله خیر است یعنی خروجی تنها دو حالت دارد که احتمال بله یا مثبت بودن پاسخ p و احتمال نه یا منفی بودن پاسخ $q= 1-p$ است.در اینجا اصطلاحا به p پارامتر توزیع گفته میشود. هر توزیع یک یا چند پارامتر توزیع دارد که با داشتن آن ها میتوان جمعیتی که از آن توزیع پیروی میکند را مدل کرد.
        <br>
        اگر X متغیر تصادفی رندوم با توزیع برنولی باشد. احتمال آن به شکل زیر تعریف می‌شود:
        \[
        Pr(X=1)=p=1-Pr(X=0)=1-q
        \]<br>
    نوع دیگری که مسئله‌های برنولی مطرح می‌شوند ترکیبی از تکرار یک کار است که هر بار آن فقط موفقیت و یا شکست دارد. برای مثال مسئله زیر را در نظر بگیرید:
    <br>
    <b>
    مسئله: 
    </b>
    یک سکه را ۵ بار پرتاب می‌کنیم به چه احتمالی (شیر، خط، شیر، شیر، خط) به ترتیب از راست به چپ ظاهر می‌شود.
    <br>
    <b>
    راه حل: 
    </b>
    در این مسئله در پرتاب اول شیر آمدن موفقیت است و در پرتاب دوم خط آمدن و به همین ترتیب. اما در همه پرتاب‌ها امکان موفقیت $\frac{1}{2}$ و امکان شکست نیز $\frac{1}{2}$ است.
    <br>
    $Pr = \frac{1}{2} \times \frac{1}{2} \times \frac{1}{2} \times \frac{1}{2} \times \frac{1}{2} = \frac{1}{2 ^ 5}$<br>
    میانگین در توزیع گسسته برنولی برابر با امید ریاضی است و به شکل زیر محاسبه می‌گردد:
    <br>
    $E[x] = Pr(X=1) \times 1 + Pr(X=0) \times 0 = p \times 1 + q \times 0 = p$<br>
    واریانس توزیع گسسته برنولی نیز به شکل زیر محاسبه می‌شود:
    <br>
    \begin{aligned}
    E[X^2]=Pr(X=1) \times 1^2 + Pr(X=0) \times 0^2 = p
     \end{aligned}
     </br>
     پس
     <br>
    \begin{aligned}
     Var = E[X^2] - E[X]^2 = p - p^2
     \end{aligned}
     </br>

<div id="sec_def_binomial" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
تعریف توزیع دو جمله ای
        </font>
		<p></p>
		<hr>
در یک آزمایش هنگامی که رخداد های مستقل  اما مشابه داریم، آن را
        <b>دنباله ای از آزمایشات مستقل</b> 
        می نامیم. وقتی که هر رخداد فقط دو حالت ممکن داشته باشد، آن را 
        <b>آزمایش برنولی</b>
        می نامیم. مثلا، در انداختن یک سکه فقط دو حالت "شیر" یا "خط" داریم.
<br>
آزمایشی را در نظر بگیرید که در آن $n$ بار یک سکه ناهمگن را می اندازیم. احتمال اینکه سکه رو بیاید برابر $p$ است، که عددی بین صفر و یک است. واضح است که این رویداد ها از هم مستقل اند. زیرا هر بار که این سکه را می اندازیم مستقل از نتایج قبل، با احتمال $p$ شیر می آید و  با احتمال $1-p$ خط.

<br>
با دانستن اینکه این رویدادها از هم مستقل اند می توانیم احتمالات را در هم ضرب کنیم و احتمال نهایی در هر کدام از حالات را بدست آوریم.  برای هر رخدادی که در آن $k$ بار شیر و $n-k$ بار خط آمده، عبارت 
$p^{k}(1-p)^{n-k}$
را میبینیم. 
<br>
حال احتمال رویداد زیر را در نظر بگیرید:
<br>
$$p(k) = P(k\:heads\:in\:n\: tosses)$$
<br>
روی دادن $k$ بار شیر در $n$ پرتاب، در هر کدام از ترتیب های آمدن  شیر و خط، یکسان است و بر اساس فرمولی که در بالا بدست آمده بدست می آید و در همه ترتیب ها این احتمال ها برابر است. کافی است تعداد حالات مختلف $k$ بار شیر آمدن را در $n$ پرتاب حساب کنیم که برابر است با:
${n \choose k}$
<br>
و فرمول کلی می شود:   
$$P(k)={n \choose k}p^{k}(1-p)^{n-k}$$
<br>
اعداد
 ${n \choose k}$
 ضرایب دوجمله ای نامیده می شوند.
<br>
متغیر تصادفی $X$ که  به صورت زیر تعریف شود از توزیع دو جمله ای پیروی می کند: 
<br>
$$ X = occurrences\: of\: event\: A\: in\: n \: trials $$
<br>
$$X \sim binomial(n,p)$$
<br>
توزیع دوجمله ای با n = 100 و p = 0.5 :
     <img src="figs/binomial.png" style="float:center; width:70%"/>
        <b>
    متغیر تصادفی دوجمله‌ای از نگاه جمع چند متغیر تصادفی برنولی:
        </b>
     هر پرتاب سکه را می‌توانیم یک توزیع برنولی در نظر بگیریم. در اینصورت متغیر تصادفی دوجمله‌ای را می‌توان جمع چند متغیر تصادفی برنولی گرفت. به طور مشخص اگر $X_1$،$X_2$،...،$X_n$ متغیرهای تصادفی برنولی مستقل از هم باشند($X_i\sim Bernoulli(p)$)، متغیر تصادفی $X=X_1+X_2+...+X_n$ توزیع دوجمله‌ای دارد($X\sim Binomial(n,p)$).
	</font>
</div>

	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
امیدریاضی و واریانس توزیع
            دوجمله ای
        </font>
		<p></p>
		<hr>
        امیدریاضی:
        <br>
        \begin{aligned} \operatorname {E}[X]=np \end{aligned}
        روش 1:
$$
\begin{aligned}
       \mu & = \operatorname{E}[X_1+X_2+...+X_n] &&&& X_i\sim Bernoulli(p)\\
           & = \operatorname{E}[X_1]+\operatorname{E}[X_2]+...+\operatorname{E}[X_n] &&&& \text{خطی بودن امید ریاضی}\\
           & = np
\end{aligned}
$$  
روش 2:
$$
\begin{aligned}
        \mu & = \sum _{k=0}^{n}k{\binom {n}{k}}p^{k}(1-p)^{n-k}\\
            & = np\sum _{k=0}^{n}k{\frac {(n-1)!}{(n-k)!k!}}p^{k-1}(1-p)^{(n-1)-(k-1)}\\
            & = np\sum _{k=1}^{n}{\frac {(n-1)!}{((n-1)-(k-1))!(k-1)!}}p^{k-1}(1-p)^{(n-1)-(k-1)}\\
            & = np\sum _{k=1}^{n}{\binom {n-1}{k-1}}p^{k-1}(1-p)^{(n-1)-(k-1)}\\
            & = np\sum _{\ell =0}^{n-1}{\binom {n-1}{\ell }}p^{\ell }(1-p)^{(n-1)-\ell }&&{\text{with }}\ell :=k-1\\
            & = np\sum _{\ell =0}^{m}{\binom {m}{\ell }}p^{\ell }(1-p)^{m-\ell }&&{\text{with }}m:=n-1\\
            & = np(p+(1-p))^{m}\\
            & = np
\end{aligned}
$$
<br>
واریانس:
<br>
$$
\begin{aligned}
       \operatorname{Var}(X) & = \operatorname{Var}[X_1+X_2+...+X_n] &&&& X_i\sim Bernoulli(p)\\
     & = \operatorname{Var}[X_1]+\operatorname{Var}[X_2]+...+\operatorname{Var}[X_n] &&&& \text{ ها مستقل از یکدیگرند$X_i$}\\
     & = np(1-p)
\end{aligned}
$$  
<br>
</font>
</div>

<div id="sec_ex_binomial" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مثال
        </font>
		<p></p>
		<hr>
        <ol>
            <li>
$X\sim Binomial(n,p)$ و $Y\sim Binomial(m,p)$ دو متغیر تصادفی مستقل از هم هستند. اگر $Z=X+Y$ باشد، PMF آن را بدست آورید.
<br/>
راه اول: X و Y را می‌توان به صورت جمع متغیرهای تصادفی برنولی مستقل از هم نوشت.
$$
X=X_1+X_2+...+X_n\\
Y=Y_1+Y_2+...+Y_m
$$
که $X_i\sim Bernoulli(p)$ ها و $Y_j\sim Bernoulli(p)$ ها مستقل از هم هستند.
در نتیجه
$$
Z=X_1+X_2+...+X_n+Y_1+Y_2+...+Y_m
$$
است. Z جمع $m+n$ متغیر تصادفی برنولی مستقل از هم است پس $Z\sim Binomial(m+n,p)$ است.
$$
\begin{aligned}
    P_{Z}(k)=\begin{cases}
        {\binom {m+n}{k}}p^{k}(1-p)^{m+n-k} & \text{k=0,1,2,...,m+n}\\
        0 & \text{otherwise}\\
    \end{cases}
\end{aligned} 
$$
                <br/>
راه دوم: برای $k\in \{ 0,1,2,...,m+n \}$ داریم
$$
\begin{aligned}
P_{Z}(k) & = P(X+Y=k)\\
         & = \sum_{i=0}^{n} P(X+Y=k|X=i)P(X=i)\\
         & = \sum_{i=0}^{n} P(Y=k-i|X=i)P(X=i)\\
         & = \sum_{i=0}^{n} P(Y=k-i)P(X=i)\\
         & = \sum_{i=0}^{min(n,k)} P(Y=k-i)P(X=i)\\
         & = \sum_{i=0}^{min(n,k)} \binom{m}{k-i} p^{k-i}(1-p)^{m-k+i}\binom{n}{i} p^{i}(1-p)^{n-i}\\
         & = \sum_{i=0}^{min(n,k)} \binom{m}{k-i} \binom{n}{i}p^{k}(1-p)^{m+n-k}\\
         & = \binom{m+n}{k} p^{k}(1-p)^{m+n-k}\\
P_{Z}(k) & = \begin{cases}
        {\binom {m+n}{k}}p^{k}(1-p)^{m+n-k} & \text{k=0, 1, 2, ..., m+n}\\
        0 & \text{otherwise}\\
    \end{cases}
\end{aligned}    
$$
            </li>
            <br/>
            <li>
اگر $X\sim Binomail(n,p)$ و $Y\sim Binomail(n-1,p)$ باشد، رابطه زیر را اثبات کنید.
$$
\operatorname{E}[X^{k}]=np\operatorname{E}[(Y+1)^{k-1}]
$$
راه حل:
$$
\begin{aligned}
\operatorname{E}[X^{k}] & = \sum_{i=0}^{n} i^{k}\binom{n}{i}p^{i}(1-p)^{n-i}\\
                        & = \sum_{i=1}^{n} i^{k-1}i\binom{n}{i}p^{i}(1-p)^{n-i}\\
                 & = \sum_{i=1}^{n} i^{k-1}n\binom{n-1}{i-1}p^{i}(1-p)^{n-i} &&& i\binom{n}{i}=n\binom{n-1}{i-1}\\
                 & = np\sum_{i=1}^{n} i^{k-1}\binom{n-1}{i-1}p^{i-1}(1-p)^{n-i}\\
                 & = np\sum_{j=0}^{n-1} (j+1)^{k-1}\binom{n-1}{j}p^{j}(1-p)^{n-1-j} &&& j=i-1\\
                 & = np\operatorname{E}[(Y+1)^{k-1}]
\end{aligned}
$$
            </li>
            <br/>
            <li>
اگر $X\sim Binomail(n,p)$ باشد، $\operatorname{Var}(X)$ را بدست آورید.
                <br/>
راه حل:
$$
\begin{aligned}
\operatorname{E}[X^{2}] & = np\operatorname{E}[(Y+1)^{1}] &&& Y\sim Binomial(n-1,p)\\
                        & = np\operatorname{E}[Y]+1\\
                        & = np((n-1)p+1)
\end{aligned}
$$
                <br/>
$\operatorname{E}[X]=np$ پس  
                <br/>
$$
\begin{aligned}
\operatorname{Var}(X) & = \operatorname{E}[X^{2}] - (\operatorname{E}[X])^{2}\\
                      & = np((n-1)p+1)-(np)^{2}\\
                      & = np(1-p)
\end{aligned}
$$            
            </li>
            <br/>
            <li>
یک بازی شامل n قسمت است. احتمال برنده شدن در هر قسمت مستقل از قسمت‌های دیگر و برابر با p است. در صورتی که حداقل برنده نصف قسمت‌ها شده باشیم برنده بازی می‌شویم. چه زمانی احتمال برنده شدن در یک بازی $2k+1$ قسمتی بیشتر از یک بازی $2k-1$ قسمتی است.
                <br/>
راه حل: متغیر تصادفی X را برابر با تعداد برنده شدن در $2k-1$ قسمت اول می‌گیریم. واقعه W، واقعه برنده شدن در بازی است.
                <br/>
برای برنده شدن در یک بازی $2k+1$ قسمتی:
                <ul>
                    <li> $X\geq k+1$ </li>
                    <li> $X=k$ و حداقل یکی از دو قسمت باقیمانده را برنده شویم. </li>
                    <li> $X=k-1$ و هر دو قسمت باقیمانده را برنده شویم. </li>
                </ul>
$$
\begin{aligned}
P_{2k+1}(W) & = P(X\geq k+1)+P(X=k)(1-(1-p)^{2})+P(X=k-1)p^{2}\\
\end{aligned}
$$
و
$$
\begin{aligned}
P_{2k-1}(W) & = P(X\geq k)\\
                                   & = P(X=k)+P(X\geq k+1)\\
\end{aligned}
$$
چون $X\sim Binomail(2k-1,p)$ است:
$$
\begin{aligned}
P_{2k+1}(W)-P_{2k-1}(W) & = P(X=k-1)p^{2}-P(X=k)(1-p)^{2}\\
     & = \binom{2k-1}{k-1} p^{k-1}(1-p)^{k}p^{2}-\binom{2k-1}{k} p^{k}(1-p)^{k-1}(1-p)^{2}\\
     & = \binom{2k-1}{k-1} p^{k}(1-p)^{k}(p-(1-p))\\
     & > 0 \Leftrightarrow p > \frac{1}{2}
\end{aligned}
$$
            </li>
        </ol>
    </font>
</div>

بازگشت به مسئله مسابقه
        </font>
		<p></p>
		<hr>
واضحا مساله قلی دو توزیع دوجمله‌ای است. در هر سوال با فرض اینکه شخص هیچ دانشی نسبت به سوال ندارد، و اینکه ۴ گزینه وجود دارد، احتمال موفقیت ۰.۲۵ است.
توزیع اول برای هر فرد و سوالاتش برقرار است. یعنی ۲۰ بار برای هر فرد آزمایش یکسانی تکرار می شود که احتمال موفقیت ۰.۲۵ است. حالا در ۲۰ سوال احتمال اینکه یک نفر بیشتر مساوی ۱۸ سوال پاسخ درست بدهد برابر است  با احتمال اینکه: دقیقا ۱۸ سوال + دقیقا ۱۹ + دقیقا ۲۰ سوال پاسخ درست بدهد.  
<br>
$$p(success\: for\: an\: individual) = p(18) + p(19) + p(20) ={20 \choose 18}(0.25)^{18}(1-0.25)^{20-18} + {20 \choose 19}(0.25)^{19}(1-0.25)^{20-19} + {20 \choose 20}(0.25)^{20}(1-0.25)^{20-20} = 1.55524 \times 10^{-9} + 5.457 \times 10^{-11} + 9.1 \times 10^{-13} = 1.61072 \times 10^{-9} $$
<br>
در مرحله بعد ۱۰ نفر داریم که هر نفر احتمال موفقیتی برابر مقدار بدست آمده در رابطه بالا دارد. بطور متوسط چقدر باید پول بدهد؟  باید امید ریاضی  توزیع جدید را بیابیم.
<br>
$$X \sim binomial(n=10,p=1.61072 \times 10^{-9})$$
<br>
$$E = np = 1.61072 \times 10^{-8}$$
<br>
این در واقع امید ریاضی تعداد افراد موفق در هر قسمت از مسابقه است. قلی به ازای هر نفر ۵ میلیون باید بدهد:
<br><br>
$$Gholi’s\:expense = E(X) * 5000000  = 0.080536\: toman$$
<br>
قلی در هر مسابقه بطور متوسط در هر قسمت 0.8 ریال  باید پول بدهد. اگرچه خیلی ارزان در می آید، اما نه کسی شرکت خواهد کرد و نه کسی آن را می بیند!
