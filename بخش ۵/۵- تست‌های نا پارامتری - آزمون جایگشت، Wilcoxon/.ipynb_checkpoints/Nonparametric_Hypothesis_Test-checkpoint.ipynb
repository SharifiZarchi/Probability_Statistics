    هر سطر مربوط به یک روز است. ستون اول متوسط بازدید از مطالب منتشر شده در آن روز و ستون دوم  فیلتر بودن یا نبودن تلگرام در آن روز را مشخص می‌کند.

```r
data=read.csv("./data/telegram_seen.csv")
data
```

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
با استفاده از داده‌هایی که داریم، میانگین بازدید از مطالب را در دوران پیش از فیلترینگ و پس از فیلترینگ محاسبه می‌کنیم و با هم مقایسه می‌کنیم:

```r
library(dplyr)
before = data %>% filter(Filtered == 'No')
after = data %>% filter(Filtered == 'Yes')
print(paste0('Average views before filtering: ',
             mean(before$Views)))
print(paste0('Average views after filtering: ',
             mean(after$Views)))
```

<br />
<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
    مشاهده می‌شود که میانگین بازدید پس از فیلترینگ کاهش یافته است. اکنون می خواهیم با استفاده از یک آزمون فرض، فرض کاهش بازدید روزانه مطالب تلگرام را بررسی کنیم.
        <br>
        شاید اولین روشی که به ذهن برسد، استفاده از t-test باشد. اما باید توجه کنیم که در آزمون تی، فرض می‌کنیم که داده‌ها از توزیع نرمال می‌آیند اما در اینجا ما نمی‌توانیم چنین فرضی راجع به داده ها داشته باشیم.
    </font>
</div>

<p></p>
<br />
<div id="sec_permutation_test" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
آزمون جایگشت (permutation test)
        </font>
		<p></p>
		<hr>
        آزمون جایگشت (permutation test)  یکی از انواع آزمون های غیر پارامتریک (non parametric) است. تفاوت این آزمون ها با آزمون های پارامتریک در این است که در اینجا فرضی راجع به توزیع متغیرها نمی‌کنیم،  اما مثلا در آزمون تی که یک آزمون پارامتریک است فرض می‌کنیم که داده از توزیع نرمال با واریانس یکسان می‌آیند.
        <br>
        <br>
        فرض کنید 2 دسته داده A و B داریم.
        دسته اول شامل $n_A$ نمونه و دسته دوم شامل $n_B$ نمونه است.
        فرض صفر این است که این دو دسته توزیع یکسانی دارند. بنابراین اگر فرض صفر  درست باشد، بین این که یک نمونه در دسته A باشد یا در دسته B، تفاوتی نیست.
        فرض کنید از آماره میانگین برای آزمون استفاده می‌کنیم. ابتدا تفاضل میانگین های دو دسته یعنی $|\mu_A - \mu_B|$ را محاسبه می‌کنیم.
        می‌خواهیم ببینیم آیا این تفاضل به اندازه کافی بزرگ هست که ما فرض صفر را رد کنیم یا خیر؟
        <br>
        داده‌های دو دسته را تجمیع می‌کنیم، یعنی تمام $n_A + n_B$ نمونه را در نظر می گیریم و به ازای همه حالت (جایگشت) های ممکن، داده ها را به دو دسته A و B به اندازه های $n_A$ و $n_B$ افراز می‌کنیم؛ یعنی تمامی حالت های نسبت دادن هر نمونه به دسته A یا B، به طوری که اندازه دسته ها به ترتیب $n_A$ و $n_B$ باشد.
        تفاضل میانگین های دو دسته را در هر حالت محاسبه می‌کنیم.
        مجموعه تمام تفاضل میانگین‌های به دست آمده، توزیع تمامی حالت‌های ممکن برای اختلاف میانگین دو دسته، به شرط درست بودن فرض صفر است.
        مقدار p-value برابر با نسبت تعداد حالت‌هایی که تفاضل میانگین بزرگتر از تفاضل میانگین اولیه است بر روی تعداد کل حالت‌ها می‌باشد.
        <br>
        دقت کنید که روش توضیح داده شده مربوط به حالتی است که فرض جایگزین ما دوطرفه باشد. اگر فرض جایگزین ما یک طرفه باشد، باید به جای $|\mu_A-\mu_B|$ از $\mu_A-\mu_B$ استفاده کنیم.
        <br>
        توجه کنید که در این آزمون تنها فرض استفاده شده این است که بتوانیم نمونه‌ها را به طور یکسان به هرکدام از دو دسته نسبت دهیم.
	</font>
</div>



<p></p>
<br />
<div id="sec_telegram_again" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
بازگشت به مسئله تلگرام
        </font>
		<p></p>
		<hr>
       آزمون جایگشت به کمک مثال روشن تر بیان می شود. قدم به قدم  آزمون جایگشت را بر روی مساله تلگرام اجرا می‌کنیم. $\alpha$ را برابر 0.05 در نظر می‌گیریم.<br>
        در مساله تلگرام، ما دو دسته داده داریم، دسته اول داده‌های مربوط به قبل فیلترینگ و دسته دوم مربوط به بعد فیلترینگ.
        ابتدا تفاضل میانگین‌های دو دسته را محاسبه می کنیم:
       </font>
</div>

```r
obsDiff = abs(mean(before$Views)-mean(after$Views))
print(paste0('Difference in Averages is: ',obsDiff))
```

<p></p>
<div id="sec_jam_again" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
اکنون تمامی حالات ممکن نسبت دادن $n_A$ نمونه به دسته A و $n_B$ نمونه به دسته B را می سازیم:
    </font>
</div>

```r
if(!require(gtools)){install.packages("gtools")}
library(gtools)
nA = 7
nB = 9
gathered = data$Views
allA = combinations(nA+nB, nA, gathered)
numPerms = dim(allA)[1]
allB = matrix(0, nrow=numPerms, ncol=nB)
for (i in 1:numPerms) {
    allB[i,] = gathered[! gathered %in% allA[i,]]
}
```

<p></p>
<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
    numPerms برابر با تعداد تمامی حالات ممکن است.
        <br>
    allA و allB دو ماتریس هستند که ابعادشان به ترتیب $numPerms \times n_A$ و $numPerms \times n_B$ است.
        به عبارتی به ازای جایگشت i ام از numPerms، $n_A$ تا نمونه در سطر iام ماتریس allA و $n_B$ نمونه بقیه در سطر iام ماتریس allB قرار گرفته است.
        مثلا در جایشگت 100ام، اعداد زیر در گروه A قرار دارند:
    </font>
</div>

```r
allA[100,]
```

<p></p>
<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
    و اعداد زیر هم در گروه B قرار دارند:
</font>
</div>

```r
allB[100,]
```

<p></p>
<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
اکنون به ازای هر جایگشت، تفاضل میانگین دو دسته را محاسبه می کنیم:
    </font>
</div>

```r
meansA = apply(allA, 1, mean)
meansB = apply(allB, 1, mean)
diffs = abs(meansA - meansB)
```

<p></p>
<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
این تفاضلات میانگین های به دست آمده به ازای هر جایگشت، در واقع توزیع اختلاف میانگین داده‌های گروه A و B با فرض درست بودن فرض صفر است. 
    اگر هیستوگرام این توزیع را رسم کنیم:</font>
</div>

```r
library(ggplot2)
ggplot(as.data.frame(diffs), aes(x=diffs)) +
 geom_histogram(binwidth = 3, fill=I('blue')) +
 geom_vline(xintercept=obsDiff, color='red', size=1.5) + 
  labs(x='Difference in averages', y='Count', title='Histogram of Difference between Averages') 

```

<p></p>
<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
در شکل بالا، خط قرمز، اختلاف میانگین‌ها در مشاهده اصلی را نشان می‌دهد. 
        p-value برابر است با احتمال اینکه اختلاف میانگین‌ها برابر یا بزرگتر از آن باشد، یعنی:
    </font>
</div>

```r
greater = diffs[diffs >= obsDiff]
pValue = length(greater) / numPerms
print(paste0("Calculated P-value is: ", pValue))
```

<p></p>
<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
    مشاهده می‌کنیم که مقدار p-value به دست آمده، کمتر از $\alpha=0.05$ است؛ بنابراین فرض صفر رد می‌شود یعنی طبق این داده‌ها، فیلترینگ تلگرام باعث کاهش بازدید از مطالب آن شده است.
        <br>
        مراحل اجرای آزمون جایگشت قدم به قدم نشان داده شد. در زبان R، می‌توان با استفاده از پکیج coin و دستور oneway_test، این آزمون را به سادگی انجام داد:

    </font>
</div>

```r
if(!require(coin)){install.packages("coin")}
library(coin)
oneway_test(Views ~ Filtered, data=data, distribution='exact')
```

<div id="sec_conc" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            آزمون Wilcoxon
        </font>
		<p></p>
		<hr>
        «خودکفایی گندم» در کشور ما مساله‌ی مهمی است
        و هر سالی که این اتفاق مبارک می‌افتد، به کشاورزان جوایز ویژه‌ای اهدا می‌شود تا در این راه تشویق شوند. اما متاسفانه با کاهش بارش‌ باران، ما هر چندسال یکبار این خودکفایی را از دست می‌دهیم.
        ،مریم و مینا، دانشجوهای زیست‌شناسی تهران و بیونفورماتیک شریف،
       پس از چندین سال کار در آزمایشگاه، شرکت «گندم» را تاسیس کردند. آن‌ها در این شرکت ترکیب ویژه‌ای از کود را تولید می‌کردند که به عقیده‌ی آن‌ها مصرف آب گندم‌ها را کاهش می‌داد.   
       آن‌ها برای این کار ده زمین زراعی را خریداری کردند.
        سپس مصرف آب هر کدام از این زمین‌ها در یک کیلومتر مربع را بررسی کردند.
        در مرحله‌ی بعدی آن‌ها کود تولیدی شرکتشان را به پای گندم‌ها ریختند و سپس بعد از یک
        هفته دوباره میزان مصرف آب هر زمین زراعی را بررسی کردند. 
        داده‌های زیر حاصل کار آن‌ها است.
        دقت کنید که ستون اول مصرف آب قبل از ریختن کود و ستون دوم مصرف آب بعد از ریختن کود است.
        مریم و مینا از شما می‌خواهند که با توجه به این داده‌ها به آن‌ها بگویید که آیا کود تولیدی آن‌ها واقعا مصرف آب گندم‌ها را کاهش داده است?
	</font>
</div>

```r
before = c(10, 20, 5, 6, 19, 16, 13, 8, 9, 11)
after= c(11, 17, 4, 8, 21, 15, 14, 9, 9, 12)
wheat <- data.frame(before = before, after = after)
wheat
```

<p></p>
<br/>
<div id="wilcoxon2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<p></p>
		<hr>
       یک شهود اولیه این است که فرض کنیم ستون 
        after 
        یک متغیر تصادفی آمده از توزیع نرمال است 
        و ستون before
        هم یک متغیر تصادفی آمده از توزیع نرمال است
        و سپس از آزمون t-test استفاده کنیم.
        با کمی بررسی موشکافانه، متوجه می‌شویم که در این روش ما از تمام آنچه که در داده پنهان است استفاده نکرده‌ایم.
         در این روش ما به هیچ‌وجه از این موضوع استفاده نکرده‌ایم که برای مثال در زمین اول میزان مصرف آب اول ۱۰ و بعد از ریختن کود ۱۱ شده است و در واقع این جفت جفت بودن اعداد هر سطر را بررسی نکرده‌ایم.
        اگر بخاطر بیاورید ما در خانواده‌ی t-test ها
        برای این مساله، تست مخصوص paired را داشتیم.
        کد این تست در زیر آمده است.
	</font>
</div>

```r
t.test(wheat$before, wheat$after, paired = T)
```

<p></p>
<br/>
<div id="wilcoxon2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<p></p>
		<hr>
       در دانش آمار به تست‌هایی که برپایه‌ی جفت بودن داده‌ها هستند، 
        تست‌های 
        <a href = "https://en.wikipedia.org/wiki/Paired_difference_test" >Paired Difference Test</a>
       می‌گویند.
         در میان تست‌های ناپارامتری تست
        Wilcoxon Signed Rank Test
        از این خانواده است و می‌تواند اینجا به کمک ما بیاید.
        بگذارید قبل از هر چیز فرض صفر و فرض یک این تست را ببینیم.
        <br>
        <ul>
            <li>
                H0:
               تفاوت بین دو توزیع از یک توزیع با امیدریاضی صفر و متقارن حول صفر پیروی می‌کند. 
            </li>
            <li>
                H1:
                تفاوت بین دو توزیع از توزیعی متقارن حول صفر پیروی نمی‌کند.
            </li>
        </ul>
        <br>
        قبل از اینکه خود تست را توضیح بدهیم، بگذارید کمی کنکاش کنیم که چرا این تست می‌تواند نتایج ارزشمندی برای ما داشته‌باشد؟
        اگر فرض
        H0
        رد شود، این به این معناست که در واقع با داده‌ای که به ما داده شده است، ما فرض 
        H0
        را به نفع 
        فرض
        H1
        رد کرده‌ایم و بنابراین یعنی ما اینکه اختلاف بین این دو داده از توزیعی متقارن حول صفر پیروی می‌کند را رد کرده‌ایم و 
       این
       حداقل به این معناست که توزیع ستون اول و دوم یکسان نیستند.
        و این نشان می‌دهد حداقل این کود بی‌اثر نبوده است و اثراتی روی میزان مصرف آب گندم‌ها داشته است.
        توجه کنید که این به هیچ‌وجه به این معنا نیست که الزاما مصرف آب کم شده است و این را باید خودمان با میانگین گرفتن از ستون اول و دوم بررسی کنیم.
        تنها چیزی که این تست به ما نشان می‌دهد این است که مثلا تفاوت مشاهده شده از لحاظ آماری significant
        هست یا خیر.
        <br>
    بگذارید ابتدا خود تست را با استفاده از R
        انجام دهیم و نتیجه‌ی آن را تفسیر کنیم. در ادامه نحوه‌ی محاسبه‌ی 
        Test Statistic
        توضیح داده شده است.
        
        </font>
</div>

```r
wilcox.test(wheat$before, wheat$after, paired = T)
```

<div id="wilcoxon2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
پس همانطور که می‌بینید فرض صفر که در واقع این است که تفاوت مصرف آب گندم‌ها پس از مصرف کود تهیه شده توسط ما از توزیعی
متقارن حول صفر پیروی می‌کند، رد نشد.
دقت داشته باشید که در اینجا این نه به معنی این است که ما صحت فرض صفر را اثبات کرده‌ایم و نه به این معنی است که فرض یک غلط است. تعبیر بسیار دقیق این است که با استفاده از داده‌ی دریافت شده، فرض صفر را نمی‌توان رد کرد.
    </font>
</div>

<p></p>
<br/>
<div id="wilcoxon2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<p></p>
		<hr>
        مراحل این تست را به ترتیب پیروی می‌کنیم. در ابتدا ستونی از قدرمطلق تفاوت بین دو ستون را درست می‌کنیم.
        در واقع ستون 
        $$dif_{i} = |after_{i} - before_{i}|$$
        را می‌سازیم.
        یک ستون دیگر که در واقع ستون علامت این تفاوت است می‌سازیم:
        $$sign_{i} = sgn(after_{i} - before_{i})$$
        دقت کنید که تابع
        sgn
        ذکر شده در واقع 
       تابع
       $$  
sgn(x) = 
     \begin{cases}
       1 & x > 0\\
       0 & \text{0}\\
       -1 & x < 0
     \end{cases}
$$
است.
        </font>
</div>

```r
library(dplyr)
wheat  %>% mutate(diff = abs(after - before)) -> wheat
wheat  %>% mutate(sign = sign(after - before)) -> wheat
wheat
```

<p></p>
<br/>
<div id="wilcoxon2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<p></p>
		<hr>
        در مرحله‌ی‌ سوم این آزمون فرض جفت‌هایی را که تفاوت آن‌ها صفر است، حذف می‌کنیم.
        و 
        $N_{r}$
        را برابر با تعداد جفت‌های باقیمانده قرار می‌دهیم.
        در داده‌ای که ما داریم، 
        $N_{r} = 9$
        است.
        </font>
</div>

```r
wheat -> wheat.test
wheat.test %>% filter(diff != 0) -> wheat.test
wheat.test %>% arrange(diff) -> wheat.test
```

<div id="wilcoxon2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
در مرحله‌ی بعدی حال جفت‌ها را بر اساس مقدار
$diff$
مرتب می‌کنیم و حالا به هر جفت رتبه‌اش در این ترتیب را به عنوان
$score$
اختصاص می‌دهیم. مقادیر مساوی میانگین رتبه‌بندی‌هایی که روی آن‌ها قرار دارند را دریافت می‌کنند.
   </font>
</div>

```r
wheat.test$rank <- seq(nrow(wheat.test))
wheat.test %>% group_by(diff) %>% summarise(R = mean(rank)) -> score.mean
wheat.test %>% full_join(score.mean) -> wheat.test
wheat.test
```

<div id="wilcoxon2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
مرحله‌ی پایانی این است که
Test Statistic 
این آزمون را محاسبه کنیم. آن را 
$W$
بنامید.
$$W = \sum_{1}^{N_{r}} sgn_{i}\times R_{i}$$
   </font>
</div>

```r
wheat.test %>% mutate(temp = sign*R)  %>% sum(.$temp)
```

<div id="wilcoxon2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<p></p>
		<hr>
ما به هر ترتیب
Test Statistic
این آزمون را محاسبه کردیم. اما مشکل اینجاست که توزیعی که 
$W$
به شرط 
$H_{0}$
از آن پیروی می‌کند، توزیعی است که بسیار پیچیده بوده و بهتر است مقادیر آن را از جداول از پیش تهیه شده ببینیم تا متوجه شویم که 
P value
تستمان چقدر بوده است. برای همین ما در متن اصلی درس، تاکید چندانی روی نحوه‌ی محاسبه‌ی این Test statistic
نداریم و بیشتر تفسیر آن برایمان اهمیت دارد.
       </font>
</div>

<div id="wilcoxon2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<p></p>
        <font color=#FF7500 size=6>
            مقایسه‌ی تست‌های پارامتریک و ناپارامتریک
        </font>
		<hr>
        دانستن کوهی از آزمون‌های فرض، هنگامی که ندانیم که در هر موقعیت از کدام آن‌ها استفاده کنیم، نه تنها مفید واقع نمی‌شود، بلکه ممکن 
است با استفاده‌ی غلط از این آزمون‌ها نتایج غلط و اشتباهی را استخراج کنیم و آن‌ها را به عنوان واقعیت در نظر بگیریم.
تست‌های ناپارامتریک با فراگیر شدن کامپیوتر‌ها امکان‌پذیر شده‌اند.
چند نکته هست که همیشه در استفاده از تست‌های ناپارامتریک باید مورد توجه ما باشند.
        <ul>
            <li>
               تست‌های ناپارامتریک معمولا وقتی مفید هستند که تعداد داده‌های ما اندک باشد. بسیاری از افراد فکر می‌کنند که تست‌های ناپارامتریک وقتی مفید هستند، که داده‌های ما از توزیع نرمال پیروی نکنند. در واقع عقیده‌ی عمومی بر این است که چون این تست‌ها توزیع نرمال را برای داده‌ها فرض نمی‌کنند، بنابراین وقتی داده‌ها از توزیع نرمال پیروی نمی‌کنند این تست‌ها انتخاب‌های بهتری هستند. اما در واقع وقتی تعداد نمونه‌های ما زیاد هستند، در بسیاری از موارد، به علت قانون حد مرکزی، تست‌های پارامتری هم بسیار دقیق عمل می‌کنند.
            </li>
            <li>
                تست‌های ناپارامتری هنگامی که تعداد داده‌ها کم هستند و از توزیع نرمال پیروی نمی‌کنند، آزمون‌های بهتری هستند.
            </li>
            <li>
                برای دیدن یک مقایسه‌ی جامع‌تر، مطالعه‌ی 
               <a href = "http://statisticsbyjim.com/hypothesis-testing/nonparametric-parametric-tests/">این</a> 
                لینک را به شما توصیه می‌کنیم.
            </li>
        </ul>
        </font>
</div>
